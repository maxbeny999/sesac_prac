
### 문제 1
사용자로부터 정수 `n`을 입력받아, 1부터 `n`까지 모든 정수의 합을 출력하세요.

````python
n = int(input())
total = 0
for num in range(n):
    total += num
print(total)
````

### 문제 3

1부터 30까지의 숫자 중에서 3의 배수를 제외한 나머지 숫자들만 출력하세요.

````python
for num in range(1, 31):
    if num % 3 != 0:
        print(num)

#예외를 처리하는 형식의 코드
for num in range(1, 31 ):
    if num % 3 == 0:
        continue
    print(num)
````

### 문제 6

사용자로부터 숫자를 계속 입력받다가, `0`을 입력하면 지금까지 입력된 모든 숫자의 합계를 출력하고 종료하는 프로그램을 작성하세요.

````python
total = 0
while True:
    number = int(input())
    if number == 0:
        break
    total += number
print(total)
````
