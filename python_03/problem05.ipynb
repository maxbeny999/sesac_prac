### 문제 11
사용자로부터 숫자를 계속 입력받다가 `-1`을 입력하면 입력을 멈추고, 그동안 입력받은 숫자들 중에서 **가장 큰 값**을 출력하는 프로그램을 `while`문을 사용하여 작성하세요. (단, `max()` 함수를 사용하지 마세요.)

```python
max_val = 0

while True:
    num = int(input("숫자를 입력하세요 (-1 종료): "))
    if num == -1:
        break
    # 여기에 최댓값을 갱신하는 로직을 작성하세요

print(f"가장 큰 값: {max_val}")
```

### 문제 12
`while`문을 사용하여 1부터 100까지의 정수 중 홀수의 합과 짝수의 합을 각각 따로 구하여 출력하세요.

```python
odd_sum = 0
even_sum = 0
n = 1

while n <= 100:
    # 조건문을 사용하여 홀수/짝수 합계를 구하세요
    n += 1

print(f"홀수의 합: {odd_sum}")
print(f"짝수의 합: {even_sum}")
```
