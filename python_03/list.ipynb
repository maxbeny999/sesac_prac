### 데이터 조회(index)

```python
names = ['alex', 'ken', 'jun']

# alex라는 데이터에 접근하고 싶어.
# print( names의 0번째 데이터를 줘)
print(names[0])

print(names[2])

name = names[2]
print(name)
```

### 데이터 수정

```python
name = 'jun' # ken으로 바꾸고 싶어.
name = 'ken' # 재할당, 덮어쓰기를 할 수 있습니다.

print(name)

names = ['alex', 'ken', 'jun']

# 'jun'이란 데이터를 'kyle'로 바꾸고 싶음.
# names의 2번째 자리에 'kyle'을 넣고 싶음.
names[2] = 'kyle'

print(names)
```
