```python
scores = [80, 100, 20, 70, 30, 10, 20]



for el in scores:
    print(el)

# 리스트의 길이를 반환하는 함수
len(scores)

for i in range(len(scores)):
    score = scores[i]

    print(i, "번째", score)
```
