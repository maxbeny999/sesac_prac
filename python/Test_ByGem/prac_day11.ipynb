### ë¬¸ì œ 1: ì¬ê³  ìì‚° ê°€ì¹˜ ê³„ì‚°
ì°½ê³ ì˜ ì¬ê³  ë°ì´í„°ì…ë‹ˆë‹¤. `category`ê°€ 'electronics'ì´ë©´ì„œ `quantity`ê°€ 1ê°œ ì´ìƒì¸ ì œí’ˆë“¤ì˜ **ì´ ìì‚° ê°€ì¹˜(price * quantityì˜ í•©)**ë¥¼ êµ¬í•˜ì„¸ìš”.

```python
inventory = [
    {'name': 'Laptop', 'category': 'electronics', 'price': 1200000, 'quantity': 5},
    {'name': 'Desk', 'category': 'furniture', 'price': 250000, 'quantity': 10},
    {'name': 'Mouse', 'category': 'electronics', 'price': 40000, 'quantity': 50},
    {'name': 'Tablet', 'category': 'electronics', 'price': 800000, 'quantity': 0},
    {'name': 'Chair', 'category': 'furniture', 'price': 150000, 'quantity': 20}
]


total_value = 0

for elements in inventory:
    if elements['category'] == 'electronics' and elements['quantity'] >= 1:
        total_value += elements['price'] * elements['quantity']


print(f"ì „ìì œí’ˆ ì¬ê³  ì´ ê°€ì¹˜: {total_value}ì›")
```

### ë¬¸ì œ 3: ìˆ˜ê°•ìƒ ê·¸ë£¹í•‘ (Grouping)
ìˆ˜ê°•ìƒ ë¦¬ìŠ¤íŠ¸ë¥¼ **ë°˜(class)** ë³„ë¡œ ê·¸ë£¹í•‘í•˜ì—¬ `class_roster` ë”•ì…”ë„ˆë¦¬ë¥¼ ë§Œë“œì„¸ìš”.
ê²°ê³¼ ì˜ˆì‹œ: `{'A': ['Alice', 'Charlie'], 'B': ['Bob'], ...}`

```python
students = [
    {'name': 'Alice', 'class': 'A'},
    {'name': 'Bob', 'class': 'B'},
    {'name': 'Charlie', 'class': 'A'},
    {'name': 'David', 'class': 'C'},
    {'name': 'Eve', 'class': 'B'}
]

class_roster = {}

for class_mates in students:
    clms = class_mates['class']
    if  clms not in class_roster:
        class_roster[clms] = []

    class_roster[clms].append(class_mates['name'])
        

print(class_roster)
```

### ë¬¸ì œ 5: ë¬¸ìì—´ íŒŒì‹±ê³¼ í†µê³„
ë¬¸ìì—´ í˜•íƒœì˜ íˆ¬í‘œ ê²°ê³¼ë¥¼ ë¶„ì„í•˜ì„¸ìš”. ê° ê³¼ì¼ì´ ëª‡ í‘œë¥¼ ë°›ì•˜ëŠ”ì§€ `vote_count` ë”•ì…”ë„ˆë¦¬ì— ì¹´ìš´íŒ…í•˜ì„¸ìš”.
ì…ë ¥ ë°ì´í„°ëŠ” ì‰¼í‘œ(`,`)ë¡œ êµ¬ë¶„ë˜ì–´ ìˆìŠµë‹ˆë‹¤.

```python
raw_votes = "apple,banana,apple,orange,banana,apple,grape,apple"

vote_count = {}

votes = raw_votes.split(",")

for fruit in votes:
    if fruit not in vote_count:
        vote_count[fruit] = 0
    
    vote_count[fruit] += 1

print(vote_count)
```
