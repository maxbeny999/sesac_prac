## 1️⃣ [복습] 카드2 (백준 2164)
- **핵심:** `collections.deque`의 `popleft()` 활용
- **조건:** N장의 카드가 있을 때, 1장 버리고 1장 뒤로 옮기기를 반복하여 마지막 남는 카드 찾기
- **링크:** [https://www.acmicpc.net/problem/2164](https://www.acmicpc.net/problem/2164)

```python
from collections import deque

def solve_card2(n):
    cards = deque(range(1, n+1))
    while len(cards) > 1:
        cards.popleft()
        move_card = cards.popleft()
        cards.append(move_card)
    return cards[0]

# 테스트
print(solve_card2(6)) # 예상 결과: 4
```

## 2️⃣ [복습] 단어 공부 (백준 1157)
- **핵심:** `upper()`로 대소문자 통일, `Dictionary`로 개수 세기
- **조건:** 가장 많이 사용된 알파벳 출력 (여러 개면 `?` 출력)
- **링크:** [https://www.acmicpc.net/problem/1157](https://www.acmicpc.net/problem/1157)

```python
def solve_word_study(word):
    counter = {}
    word = word.upper()
    for alphabets in word:
        counter[alphabets] = counter.get(alphabets, 0) + 1
    max_val = max(counter.values())
    winners = [key for key, value in counter.items() if value == max_val]
    if len(winners) > 1:
        return "?"
    else:
        return winners[0]

# 테스트
print(solve_word_study("Mississipi")) # 예상 결과: ?
print(solve_word_study("zZa"))        # 예상 결과: Z
```

## 3️⃣ [New] 베스트셀러 (백준 1302)
- **유형:** 해시맵 (Dictionary)
- **목표:** 오늘 가장 많이 팔린 책 제목 찾기
- **조건:** 가장 많이 팔린 책이 여러 개라면, **사전 순으로 가장 앞선 제목**을 출력하세요.
- **힌트:** `dict`로 카운팅 후 `items()`를 정렬하거나, 최대값을 찾으세요.
- **링크:** [https://www.acmicpc.net/problem/1302](https://www.acmicpc.net/problem/1302)

```python
# 입력 예시
# 5
# top
# top
# top
# top
# kimtop

import sys
# input = sys.stdin.readline # 백준 제출 시 주석 해제

# 여기에 풀이를 작성하세요

books = ["top", "top", "top", "top", "kimtop"]

def solve_bestseller(books):
    # 1. 팔린 개수 세기 (dict)
    counter = {}
    for book in books:
        counter[book] = counter.get(book, 0) + 1
    
    # 2. 가장 많이 팔린 숫자 찾기 (max)
    max_book = max(counter.values())

    # 3. 우승 후보들 리스트에 담기 (List Comprehension)
    candidate = [key for key, value in counter.items() if value == max_book ]

    # 4. 정렬하고 1등 뽑기 (sort)
    candidate.sort()

    return candidate [0]

print(solve_bestseller(books))
```

## 4️⃣ [New] 크로아티아 알파벳 (백준 2941)

```python
croatia_alphabets = ['c=', 'c-', 'dz=', 'd-', 'lj', 'nj', 's=', 'z=']
word = input() # 예: ljes=njak

# 여기에 풀이를 작성하세요

for words in croatia_alphabets:
    word = word.replace(words, "*")
print(len(word))
print(word)

```
