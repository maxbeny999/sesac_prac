### ë¬¸ì œ 8: ë¹„ë™ê¸° í•¨ìˆ˜ ë³‘ë ¬ ì‹¤í–‰ (Async/Await)

1. `fetch_user_data()`: 2ì´ˆ ì†Œìš” (asyncio.sleep(2) í™œìš©)

*íŒíŠ¸: `asyncio.gather()`ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.*

```python
import asyncio
import time

async def fetch_user_data():
    print("ìœ ì € ë°ì´í„° ì¡°íšŒ ì‹œì‘...")
    await asyncio.sleep(2)
    print("ìœ ì € ë°ì´í„° ì¡°íšŒ ì™„ë£Œ!")
    return "User Info"

async def fetch_order_data():
    print("ì£¼ë¬¸ ë°ì´í„° ì¡°íšŒ ì‹œì‘...")
    await asyncio.sleep(1)
    print("ì£¼ë¬¸ ë°ì´í„° ì¡°íšŒ ì™„ë£Œ!")
    return "Order Info"

async def main():
    start = time.time()
    
    await asyncio.gather(fetch_user_data(), fetch_order_data())
    
    end = time.time()
    print(f"ì´ ì†Œìš” ì‹œê°„: {end - start:.2f}ì´ˆ")

# ì‹¤í–‰ ì½”ë“œ (Jupyter í™˜ê²½)
await main()
```

### ë¬¸ì œ 9: ì˜ì¡´ì„± ì£¼ì… (Dependency Injection) íŒ¨í„´

```python
class EmailSender:
    def send(self, message):
        print(f"[ì´ë©”ì¼ ì „ì†¡] {message}")

class NotificationService:
    def __init__(self, sender):
        self.sender = sender

    
    def notify(self, message):
        self.sender.send(message)

# í…ŒìŠ¤íŠ¸ ì½”ë“œ
sender_instance = EmailSender()
service = NotificationService(sender_instance) # ê°ì²´ë¥¼ ë°–ì—ì„œ ë§Œë“¤ì–´ì„œ ë„£ì–´ì¤Œ (í•µì‹¬)
service.notify("íšŒì›ê°€ì…ì„ í™˜ì˜í•©ë‹ˆë‹¤!")
```
