### 문제 4: 성적별 학생 그룹핑 (Dictionary Grouping)
**목표:**
1. `students` 딕셔너리에는 학생 이름과 점수가 들어있습니다.
2. 점수에 따라 등급을 나누고, **같은 등급인 학생들의 이름을 리스트로 묶어서** 새로운 딕셔너리 `grade_group`을 만드세요.

**기준:**
- 90점 이상: "A"
- 80점 이상: "B"
- 나머지: "C"

**예상 출력:**
`{'A': ['Alice', 'David'], 'B': ['Eve', 'Bob'], 'C': ['Charlie']}`

```python
# [문제 4 작성 셀]
students = {
    "Alice": 95,
    "Bob": 82,
    "Charlie": 60,
    "David": 90,
    "Eve": 88
}
# for 문으로 딕셔너리.아이템 으로 키, 벨류 둘다 꺼낸다.
# 꺼내고 벨류 값에 따라 등급별로 새로운 딕셔너리에 넣는다.
# 
grade_group = {}
# from collections import defaultdict
# grade_group = defaultdict(list)

for student, grade in students.items():
    if grade >= 90:
        target_key = 'A'
    elif grade >= 80 and grade < 90:
        target_key = 'B'
    else:
        target_key = 'C'
    if target_key not in grade_group:
        grade_group[target_key] = []

    grade_group[target_key].append(student)
print(grade_group)
```

### 문제 5: 딕셔너리 뒤집기 (Value to Key)
**목표:**
1. `playlist` 딕셔너리에는 `노래제목: 가수이름`이 들어있습니다.
2. 이를 뒤집어서 `가수이름: [노래목록]` 형태의 새로운 딕셔너리 `artist_songs`를 만드세요.
3. 한 가수가 여러 노래를 불렀다면 리스트에 모두 추가되어야 합니다.

**예상 출력:**
`{'IU': ['Blueming', 'Love poem'], 'BTS': ['Dynamite', 'Butter'], 'Zico': ['Any song']}`

```python
# [문제 5 작성 셀]
playlist = {
    "Blueming": "IU",
    "Dynamite": "BTS",
    "Love poem": "IU",
    "Butter": "BTS",
    "Any song": "Zico"
}


artist_songs = {}
for songs, artist in playlist.items():
    if artist not in artist_songs:
        artist_songs[artist] = []
    artist_songs[artist].append(songs)

print(artist_songs)
```

# [Part 3] 알고리즘 챌린지 (Algorithm)

### ⚔️ 알고리즘 1: 괄호 짝 맞추기 (Valid Parentheses)
**설명:**
여는 괄호 `(` 와 닫는 괄호 `)` 로만 이루어진 문자열이 주어집니다.
괄호의 짝이 올바르게 맞으면 `True`, 아니면 `False`를 반환하는 함수를 작성하세요.

**조건:**
1. `(())` -> True
2. `(()` -> False (닫는 괄호 부족)
3. `)()` -> False (닫는 괄호가 먼저 나옴)
4. `(()))` -> False (닫는 괄호가 너무 많음)

```python
# [알고리즘 1 작성 셀]
# (가 들어오면 그짝인 )도 반드시 들어와야 한다.
# (가 들어온 만큼 )가 들어와하고 그 )부터 들어온건 의미 없음. )(을 거르기 위함
# 들어온 내용물을 다 받아서 딕셔너리에 넣고 (가 들어온 만큼 )가 들어오면 True 아니면 False로 간다

def check_parenthesis(s):

    stack = []

    for char in s:
        if char == "(":
            stack.append(char)
        elif char ==")":
            if not stack:
                return False
            stack.pop()
    return len(stack) == 0

print(check_parenthesis("(())"))  # True
print(check_parenthesis("(()"))   # False
print(check_parenthesis(")("))    # False
print(check_parenthesis("(()))")) # False
```

### ⚔️ 알고리즘 2: 아나그램 판별 (Anagram)
**설명:**
두 문자열 `word1`과 `word2`가 주어집니다.
두 단어가 **사용된 알파벳의 종류와 개수가 완벽히 일치**하는지 확인하는 함수를 만드세요.
(순서는 상관없습니다.)

**조건:**
1. "listen", "silent" -> True
2. "apple", "papel" -> False (p의 개수가 다름)

```python
# [알고리즘 2 작성 셀]
# 들어오는 word1를 분해하고 순서에 상관 없이 그 알파벳이 몇개 들어있는지만 확인한다.
#  그리고 그 word2도 분해 한 후 알파벳 개수가 똑같으면 정답
# ex apple a=1 p=2 l=1 e=1 papel p=2 a=1 e=1 l=1
# for문으로 word1, word2를 분해하고 리스트 그 글자가 몇개 나왔는지 세기 위해 리스트에 넣는다?
def is_anagram(word1, word2):

    count_box1 = {}
    count_box2 = {}
    for char in word1:
        if char not in count_box1:
            count_box1[char] = 1
        else:
            count_box1[char] +=1

    for char in word2:
        if char not in count_box2:
            count_box2[char] = 1
        else:
            count_box2[char] +=1
    return count_box1 == count_box2
    
print(is_anagram("listen", "silent")) 
print(is_anagram("race", "care"))     
print(is_anagram("apple", "papel"))  
```
