### Q1. 신호등 판독기 (If문)
**문제:** 변수 `color`에 신호등 색상이 들어옵니다. (`"Green"`, `"Yellow"`, `"Red"`)
1. `"Green"`이면 "건너세요" 출력
2. `"Yellow"`면 "서행하세요" 출력
3. `"Red"`면 "멈추세요" 출력
4. 그 외의 색이면 "고장입니다"를 출력하세요.

```python
color = "Red"

if color == "Green":
    print("건너세요")
elif color == "Yellow":
    print("서행하세요")
elif color == "Red":
    print("멈추세요")
else:
    print("고장입니다")
```

### Q2. 짝수만 더하기 (For + If)

### Q3. 비밀번호 맞추기 (While + Break)
**문제:** `correct_pw`는 "1234"입니다. 사용자에게 비밀번호를 계속 입력받으세요.
1. 입력받은 비밀번호가 틀리면 "틀렸습니다"를 출력하고 다시 입력받습니다.
2. 맞으면 "성공!"을 출력하고 반복을 종료(`break`)하세요.
(무한 루프 `while True` 사용)

```python
correct_pw = "1234"

while True:
    pw = input()
    if pw != "1234":
        print("틀렸습니다")
    elif pw == correct_pw:
        print("성공!")
        break
```

### Q6. 최대값 찾기 함수 (Function)
**문제:** 숫자 리스트를 입력받아 그 중 **가장 큰 수**를 반환(`return`)하는 함수 `find_max_number`를 작성하세요.
(파이썬 내장 함수 `max()`를 쓰지 말고, `for`문과 `if`문을 이용해 직접 구현해보세요.)

```python
def find_max_number(nums):
    max_num = 0
    for number in nums:
        if number > max_num:
            max_num = number
    return max_num

print(find_max_number([10, 5, 30, 20])) # 결과: 30
```

### Q9. 단어 빈도수 세기 (Dict Logic - Day 5 복습)
**문제:** `data` 리스트에 과일 이름들이 섞여 있습니다. 각 과일이 몇 개씩 나왔는지 `count_dict`에 정리하세요.
**힌트:** `if not in` 패턴을 기억하세요.

```python
data = ["Apple", "Banana", "Apple", "Cherry", "Banana", "Apple"]
count_dict = {}

for fruits in data:
    if fruits not in count_dict:
        count_dict[fruits] = 1
    elif fruits in count_dict:
        count_dict[fruits] +=1

print(count_dict) # 예상: {'Apple': 3, 'Banana': 2, 'Cherry': 1}
```

### Q10. 짝수/홀수 그룹핑 (Advanced Logic - Day 5 복습)
**문제:** `numbers`의 숫자들을 **"Even"(짝수)**과 **"Odd"(홀수)** 그룹으로 나누어 딕셔너리 `group`에 저장하세요.
**목표:** `{'Odd': [1, 3, 5], 'Even': [2, 4]}` 형태가 되어야 합니다.

```python
numbers = [1, 2, 3, 4, 5]
group = {}
# for 문으로 numbers 해체
# if 문으로 짝순지 홀순지 나누고 결과에 맞춰 리스트에 저장.
# group에 키값 벨류갑(리스트) 으로 넣음
even = []
odd = []
for number in numbers:
    if number % 2 == 0:
        even.append(number)
        group["Even"] = even
    else:
        odd.append(number)
        group["Odd"] = odd


print(group)

numbers = [1, 2, 3, 4, 5]
group = {}

for number in numbers:
    # 1. 어떤 그룹인지 이름표(Key)를 먼저 정합니다.
    if number % 2 == 0:
        key = "Even"
    else:
        key = "Odd"
    
    # 2. 방이 없으면 만듭니다. (Dynamic Grouping)
    if key not in group:
        group[key] = []
    
    # 3. 그 방에 숫자를 넣습니다.
    group[key].append(number)

print(group)
# 출력: {'Odd': [1, 3, 5], 'Even': [2, 4]}
```
