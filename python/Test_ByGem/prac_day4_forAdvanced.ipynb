## 문제 1: 소수(Prime Number) 찾기
**목표:**
1. 2부터 50까지의 숫자 중에서 **소수**만 찾아서 `primes` 리스트에 담으세요.
2. **소수(Prime Number)**란? 1과 자기 자신으로만 나누어떨어지는 수 (예: 2, 3, 5, 7, 11...)

**힌트:**
- 이중 `for`문이 필요합니다.
- 바깥 루프(`num`): 2~50 반복 (검사 대상)
- 안쪽 루프(`i`): 2부터 `num-1`까지 반복하며 나누어떨어지는지 확인 (`if num % i == 0`)
- 만약 중간에 한 번이라도 나누어떨어지면? 소수가 아님! (`break`)

````python
# [문제 1 작성 셀]
primes = []
is_primes = True

for num in range(2,51):
    is_primes = True
    for i in range(2,num-1):
        if num % i == 0:
            is_primes = False
            break
    if is_primes == True:
        primes.append(num)



print(f"2~50 사이 소수: {primes}")
````
