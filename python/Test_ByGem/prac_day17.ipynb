### Q3. 안전한 데이터 조회 (Read & `.get()`)
딕셔너리에서 존재하지 않는 키를 대괄호 `[]`로 찾으려 하면 `KeyError`가 발생하며 프로그램이 멈춥니다.
에러 없이 안전하게 데이터를 가져오는 `.get()` 메서드를 사용하여, `"email"` 키의 값을 가져오되, 만약 키가 없다면 `"정보 없음"` 이라는 기본 문자열을 반환하도록 코드를 작성하세요.

```python
user_info = {'age': 26}

# email 변수에 .get()을 활용한 결과를 저장하세요.
email = user_info.get("email", "정보없음")
# 테스트
print("Q3 결과:", email)
# 예상 출력: 정보 없음
```

### Q4. 딕셔너리 전체 순회 (`.items()`)
딕셔너리의 모든 키와 값을 한 번에 순회하려면 `.items()`를 사용합니다.
주어진 `inventory` 딕셔너리를 순회하면서, 재고(value)가 0인 과일(key)의 이름을 리스트에 담아 반환하는 함수를 완성하세요.

```python
def find_empty_stock(inventory: dict) -> list:
    empty_items = []
    
    for key, val in inventory.items():
        if val == 0:
            empty_items.append(key)
            
    return empty_items

# 테스트
print("Q4 품절 상품:", find_empty_stock({'Apple': 10, 'Banana': 0, 'Melon': 5, 'Orange': 0}))
# 예상 출력: ['Banana', 'Orange']
```

### Q5. 예외 처리 기초 (`try-except`)
서버는 어떤 상황에서도 죽으면 안 됩니다. 숫자를 0으로 나누려고 할 때 발생하는 `ZeroDivisionError`를 방어해 보겠습니다.
`try` 블록 안에서 `a / b`를 실행하고, 에러가 발생하면 `except ZeroDivisionError:` 블록에서 `"0으로 나눌 수 없습니다"` 문자열을 반환하도록 함수를 완성하세요.

```python
def safe_divide(a: int, b: int):
    try:
        return a / b
    except ZeroDivisionError:
        return ("0으로 나눌 수 없습니다")
           
# 테스트
print("Q5 정상 계산:", safe_divide(10, 2))
print("Q5 에러 방어:", safe_divide(10, 0))
# 예상 출력: 
# 5.0
# 0으로 나눌 수 없습니다
```

### Q6. 딕셔너리 예외 처리 (`KeyError`)
서버로 들어온 데이터(`payload`)에서 `"password"` 키를 조회하려고 합니다.
대괄호 `[]`를 사용하여 접근하되, 키가 없어서 `KeyError`가 발생하면 `except KeyError:` 블록에서 `False`를 반환하고, 성공적으로 찾았다면 `True`를 반환하세요.

```python
def has_password(payload: dict) -> bool:
    try:
        payload["password"]
        return True
    except KeyError:
        return False
    

# 테스트
print("Q6 패스워드 있음:", has_password({"username": "Alice", "password": "1234"}))
print("Q6 패스워드 없음:", has_password({"username": "Bob"}))
# 예상 출력: 
# True
# False
```
