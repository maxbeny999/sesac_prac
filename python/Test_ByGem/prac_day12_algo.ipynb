### ë¬¸ì œ 10: ë±(Deque)ìœ¼ë¡œ ì§„ì§œ í êµ¬í˜„í•˜ê¸° (Bronze)
ë¦¬ìŠ¤íŠ¸ì˜ `pop(0)`ì€ ì†ë„ê°€ ëŠë¦½ë‹ˆë‹¤(ë°ì´í„°ê°€ ë§ìœ¼ë©´ ë°€ë ¤ë‚˜ëŠë¼ ì˜¤ë˜ ê±¸ë¦¼).
ê·¸ë˜ì„œ ì‹¤ë¬´ì™€ ì½”ë”©í…ŒìŠ¤íŠ¸ì—ì„œëŠ” `collections.deque`ë¥¼ ì”ë‹ˆë‹¤.

1. `deque`ë¥¼ ì‚¬ìš©í•´ ë¹ˆ íë¥¼ ë§Œë“œì„¸ìš”.
2. `append()`ë¡œ 10, 20, 30ì„ ìˆœì„œëŒ€ë¡œ ë„£ìœ¼ì„¸ìš”.
3. `popleft()`ë¡œ ëª¨ë‘ êº¼ë‚´ì„œ ì¶œë ¥í•˜ì„¸ìš”. (`pop(0)` ëŒ€ì‹  `popleft`ë¥¼ ì”ë‹ˆë‹¤)

```python
from collections import deque

queue = deque()

# 1. ë°ì´í„° ë„£ê¸° (10, 20, 30)
queue.extend([10, 20, 30])
# 2. ë°˜ë³µë¬¸ìœ¼ë¡œ ëª¨ë‘ êº¼ë‚´ê¸° (popleft)
while len(queue) > 0:
    poped = queue.popleft()
    print(poped)
```

### ğŸ”¥ Warm-up 11: í•œ ë²ˆë§Œ ë’¤ë¡œ ë³´ë‚´ê¸° (Rotation)
ì¹´ë“œ ì„ê¸° ë¬¸ì œì˜ í•µì‹¬ ë¡œì§ì…ë‹ˆë‹¤.
**"ë§¨ ì•ì„ ë½‘ì•„ì„œ, ë§¨ ë’¤ë¡œ ë¶™ì¸ë‹¤"** ì´ ë™ì‘ì„ ë”± í•œ ë²ˆë§Œ ìˆ˜í–‰í•´ë³´ì„¸ìš”.

1. `deque([1, 2, 3, 4])`ê°€ ìˆìŠµë‹ˆë‹¤.
2. 1ì„ ë½‘ì•„ì„œ(`popleft`), ë’¤ì— ë¶™ì´ì„¸ìš”(`append`).
3. ê²°ê³¼ê°€ `[2, 3, 4, 1]`ì´ ë˜ë©´ ì„±ê³µì…ë‹ˆë‹¤.

```python
from collections import deque
cards = deque([1, 2, 3, 4])

# 1. ë§¨ ì• ë½‘ê¸°
top = cards.popleft()

# 2. ë§¨ ë’¤ì— ë¶™ì´ê¸°
cards.append(top)

print(cards) # deque([2, 3, 4, 1]) ì´ì–´ì•¼ í•¨
```

### ğŸ”¥ Warm-up 12: Kë§Œí¼ ê±´ë„ˆë›°ê¸° (rotate ë©”ì„œë“œ)

* `q.rotate(-2)`: ì™¼ìª½ìœ¼ë¡œ 2ì¹¸ íšŒì „ (ì• 2ëª…ì´ ë§¨ ë’¤ë¡œ ê°)

### ë¬¸ì œ 12: ìš”ì„¸í‘¸ìŠ¤ ìˆœì—´ (Silver / ë°±ì¤€ 11866)
Nëª…ì˜ ì‚¬ëŒ ì¤‘ Kë²ˆì§¸ ì‚¬ëŒì„ ê³„ì† ì œê±°í•©ë‹ˆë‹¤.
**[Logic]**
1. ì œê±°í•  ì‚¬ëŒ ì•ê¹Œì§€(`K-1`ëª…)ëŠ” **íšŒì „ì‹œì¼œì„œ ë’¤ë¡œ ë³´ëƒ…ë‹ˆë‹¤.** (`rotate(-(K-1))`)
2. ì´ì œ ë§¨ ì•ì— ì˜¨ ì‚¬ëŒì´ ì œê±° ëŒ€ìƒì…ë‹ˆë‹¤. **êº¼ëƒ…ë‹ˆë‹¤.** (`popleft`)
3. ì´ê±¸ ì‚¬ëŒì´ ë‹¤ ì—†ì–´ì§ˆ ë•Œê¹Œì§€ ë°˜ë³µí•©ë‹ˆë‹¤.

N=7, K=3ì¼ ë•Œ ì œê±° ìˆœì„œë¥¼ êµ¬í•˜ì„¸ìš”.

```python
N, K = 7, 3
people = deque(range(1, N + 1))
result = []

while people:
    # 1. K-1ë§Œí¼ ì™¼ìª½ìœ¼ë¡œ íšŒì „ (ì‚¬ëŒë“¤ì„ ë’¤ë¡œ ë³´ëƒ„)
    people.rotate(-(K-1))
    # 2. ë§¨ ì•ì‚¬ëŒ ì œê±° ë° ì €ì¥
    num = people.popleft()
    result.append(num)
print(result)
```
