 #  문제 1: 문자열 쪼개고 합치기 (Split & Join)
  **목표:**
 1. `raw_date` 문자열은 "년:월:일" 형식입니다.
 2. 이를 `split`을 사용해 쪼갠 뒤, 다시 `join`을 사용해 "년-월-일" 형식으로 바꾸세요.

  **힌트:**
 - `.split(":")` -> 리스트가 됩니다.
 - `"-".join(리스트)` -> 문자열이 됩니다.

```python
# [문제 1 작성 셀]
raw_date = "2024:01:22"
formatted_date = ""
date = raw_date.split(":")
print("-".join(date))
```

 ## 문제 4: 리스트 중복 한 방에 제거 (Set Conversion)

 ## 문제 5: 교집합 구하기 (Set Operation)
 **목표:**
 1. 두 사람(`person_a`, `person_b`)이 **공통으로 좋아하는 취미**를 찾으세요.
 2. 집합의 교집합 연산자(`&`)를 사용하세요.

 **힌트:**
 - 집합끼리는 `&`(교집합), `|`(합집합), `-`(차집합) 연산이 가능합니다.

```python
person_a = {"축구", "야구", "농구"}
person_b = {"농구", "배구", "축구"}
common_hobbies = set(person_a & person_b)
print(common_hobbies)
```
