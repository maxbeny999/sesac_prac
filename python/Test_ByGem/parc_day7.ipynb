### 문제 3: 행동 부여하기 (Method)
**목표:** 객체가 인사를 할 수 있도록 메서드를 추가합니다.

**조건:**
1. 위에서 만든 `Person` 클래스에 `introduce` 메서드를 추가하세요.
2. 이 메서드는 "안녕하세요, 제 이름은 [이름]이고 [나이]살입니다."를 출력해야 합니다.
3. 객체를 생성하고 `introduce()`를 호출하세요.

```python
# [문제 3 작성 셀]
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"안녕하세요, 제 이름은 {self.name}이고 {self.age}살입니다.")
        
p1 = Person("Alice", 30)

p1.introduce()
```

### 문제 4: 상태 변경하기
**목표:** 메서드를 통해 객체의 속성(나이)을 변경해봅니다.

**조건:**
1. `Person` 클래스에 `birthday` 메서드를 추가하세요.
2. 이 메서드가 호출되면 `age`가 1 증가하고, "생일 축하합니다!"를 출력합니다.
3. 객체의 나이를 확인하고, `birthday()` 호출 후 다시 나이를 확인하세요.

```python
# [문제 4 작성 셀]
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"안녕하세요, 제 이름은 {self.name}이고 {self.age}살입니다.")

    def birthday(self):
        self.age += 1
        print("생일 축하합니다!")
    
p1 = Person("Alice", 30)

p1.introduce()
p1.birthday()

p1.introduce()
```

### 문제 6: 쇼핑 카트 (List Attribute)
**목표:** 클래스 속성으로 **리스트**를 사용하여 여러 데이터를 관리합니다.

**조건:**
1. `Cart` 클래스를 만드세요.
2. `__init__`에서 빈 리스트 `self.items`를 생성하세요.
3. `add_item(item)`: 물건 이름을 받아 리스트에 추가합니다.
4. `show_items()`: 현재 담긴 모든 물건을 리스트 형태로 출력합니다.

```python
# [문제 6 작성 셀]
class Cart:
    # [Step 1] 생성자: 카트를 처음 뽑을 때 딱 1번 실행됩니다.
    def __init__(self):
        # 1-1. 'items'라는 이름의 빈 리스트(창고)를 내 몸(self)에 만듭니다.
        # 이 리스트는 객체가 살아있는 동안 계속 유지됩니다.
        self.items = [] 

    # [Step 2] 물건 담기: 외부에서 item을 던져주면 실행됩니다.
    # 예: cart.add_item("사과") 라고 부르면 -> item은 "사과"가 됩니다.
    def add_item(self, item):
        # 2-1. 내 창고(self.items)에 받아온 물건(item)을 추가(append)합니다.
        # 굳이 self.item = item 처럼 따로 저장할 필요 없이, 바로 넣고 끝냅니다.
        self.items.append(item) 

    # [Step 3] 확인하기: 현재 상태를 보고 싶을 때 실행합니다.
    def show_items(self):
        # 3-1. 창고(self.items)를 통째로 돌려줍니다.
        return self.items

# --- 실제 사용 예시 (실행 흐름) ---
my_cart = Cart()          # [Step 1] 실행 -> 빈 리스트 [] 생성됨
my_cart.add_item("사과")   # [Step 2] 실행 -> 리스트가 ["사과"]가 됨
my_cart.add_item("바나나") # [Step 2] 실행 -> 리스트가 ["사과", "바나나"]가 됨
print(my_cart.show_items()) # [Step 3] 실행 -> ["사과", "바나나"] 출력
```
