### [Warm-up 1] 딕셔너리 방 확인하기

**조건:**
1. `inventory`에 "orange"라는 키가 **없다면**, 값을 0으로 새로 만드세요.
2. 그 후, "orange"의 개수를 3개 증가시키세요.

```python
inventory = {"apple": 5, "banana": 2}
target_item = "orange"

# 1. 있는지 확인하고 없으면 0으로 만들기

# 2. 개수 증가시키기

if "orange" not in inventory:
    inventory['orange'] = 0
inventory[target_item] +=3
print(inventory)
```

### [Warm-up 3] 빈 방 만들기 패턴 (Not In)
**목표:** 리스트를 순회하며 딕셔너리에 **'방(List)'이 없을 때만 생성**하는 패턴을 익힙니다.

3. (이 문제는 데이터를 채우지는 않고, 방을 만드는 것까지만 합니다.)

```python
names = ["Tom", "Jane", "Tom", "Max"]
name_dict = {}

for name in names:
    if name not in name_dict:
        name_dict[name] = []

    pass

print(name_dict) 
# 예상 결과: {'Tom': [], 'Jane': [], 'Max': []}
```

### [Warm-up 4] 분류 로직 맛보기
**목표:** 숫자를 판단하여 그룹(Key) 이름을 정하는 로직을 연습합니다.

**조건:**
1. `age` 변수의 값에 따라 `group_key` 변수에 문자열을 저장하세요.
2. 20 미만이면 "Teen", 20 이상 30 미만이면 "20s", 30 이상이면 "30s".

```python
ages = [15, 25, 33]

for age in ages:
    group_key = ""
    if age < 20:
        group_key = "Teen"
    elif age < 30:
        group_key = "20s"
    elif age >= 30 :
        group_key = "30s"
    
    print(f"{age}살은 {group_key} 그룹입니다.")
```

### [Advanced 1] 경력별 사원 그룹핑 (Grouping)
**목표:** `employees` 딕셔너리(사원명: 경력연수)를 **경력 수준별로 묶어서** 새로운 딕셔너리 `career_group`을 만드세요.

**분류 기준:**
- 5년 미만: "Junior"
- 5년 이상 10년 미만: "Senior"
- 10년 이상: "Expert"

**예상 출력:**
`{'Senior': ['Alice', 'David'], 'Junior': ['Bob'], 'Expert': ['Charlie']}`

```python
employees = {
    "Alice": 7,
    "Bob": 3,
    "Charlie": 15,
    "David": 5
}

career_group = {}

for employee, years in employees.items():
    if years < 5:
        grade = "Junior"
    elif years < 10:
        grade = "Senior"
    elif years >= 10:
        grade = "Expert"
    if grade not in career_group:
        career_group[grade] = []
    career_group[grade].append(employee)
    

print(career_group)
```

### [Advanced 2] 국가별 도시 정리 (Inverting)
**목표:** `city_country` (도시: 국가) 딕셔너리를 뒤집어서 `country_cities` (국가: [도시 목록]) 형태로 만드세요.

**조건:**
1. 한 국가에 여러 도시가 있다면 리스트에 모두 담겨야 합니다.
2. 딕셔너리에 방이 없으면 만들고(`[]`), 있으면 추가하는(`append`) 로직을 활용하세요.

**예상 출력:**
`{'Korea': ['Seoul', 'Busan'], 'Japan': ['Tokyo'], 'USA': ['New York']}`

```python
city_country = {
    "Seoul": "Korea",
    "Tokyo": "Japan",
    "Busan": "Korea",
    "New York": "USA"
}

country_cities = {}

for city, country in city_country.items():
    if country not in country_cities:
        country_cities[country] = []
    country_cities[country].append(city)
    
print(country_cities)
```
