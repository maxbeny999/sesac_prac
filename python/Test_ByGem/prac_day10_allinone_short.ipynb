## ğŸ”´ Advanced: ë°ì´í„° ê´€ë¦¬ì (Class & Logic)

### [Advanced 1] ë‚˜ë§Œì˜ ì¼ê¸°ì¥ (`Diary` í´ë˜ìŠ¤)
ë§¤ì¼ë§¤ì¼ ì¼ê¸°ë¥¼ íŒŒì¼ì— ì €ì¥í•´ì£¼ëŠ” í´ë˜ìŠ¤ë¥¼ ë§Œë“­ë‹ˆë‹¤.

**ë¯¸ì…˜:**
`Diary` í´ë˜ìŠ¤ë¥¼ ì™„ì„±í•˜ì„¸ìš”.
1. **`__init__(self, filename)`:** ì €ì¥í•  íŒŒì¼ ì´ë¦„(ì˜ˆ: `diary.txt`)ì„ ë°›ì•„ì„œ ì €ì¥í•©ë‹ˆë‹¤.
2. **`write_entry(self, text)`:**
    * `text`(ì¼ê¸° ë‚´ìš©)ë¥¼ ì…ë ¥ë°›ìŠµë‹ˆë‹¤.
    * íŒŒì¼ì— **ì´ì–´ì“°ê¸° ëª¨ë“œ `'a'` (Append)** ë¡œ ì—½ë‹ˆë‹¤. (ë®ì–´ì“°ë©´ ì˜›ë‚  ì¼ê¸°ê°€ ë‚ ì•„ê°€ë‹ˆê¹Œìš”!)
    * ë‚´ìš©ê³¼ í•¨ê»˜ **ì¤„ë°”ê¿ˆ(`\n`)**ì„ ì¶”ê°€í•´ì„œ ì €ì¥í•˜ê³  ë‹«ìŠµë‹ˆë‹¤.

*(í…ŒìŠ¤íŠ¸ í•  ë•Œ ì—¬ëŸ¬ ë²ˆ ì‹¤í–‰í•´ì„œ ë‚´ìš©ì´ ê³„ì† ì¶”ê°€ë˜ëŠ”ì§€ í™•ì¸í•´ë³´ì„¸ìš”!)*

```python
# [Advanced 1 ì‘ì„± ì…€]
class Diary:
    def __init__(self, filename):
        self.filename = filename

    def write_entry(self, text):
        f = open(self.filename, "a", encoding='utf-8' )
        f.write(text + "\n")
        f.close()

# í…ŒìŠ¤íŠ¸
my_diary = Diary("my_diary.txt")
my_diary.write_entry("ì˜¤ëŠ˜ íŒŒì´ì¬ ê³µë¶€ê°€ ì¬ë¯¸ìˆì—ˆë‹¤.")
my_diary.write_entry("íŒŒì¼ ì…ì¶œë ¥ë„ ì´ì œ í•  ì¤„ ì•ˆë‹¤!")

# í™•ì¸ìš© ì½”ë“œ (íŒŒì¼ ì½ì–´ì„œ ì¶œë ¥)
print(open("my_diary.txt", "r", encoding='utf-8').read())
```

### [Advanced 2] ì„±ì¸ ì¸ì¦ ì‹œìŠ¤í…œ (ë”•ì…”ë„ˆë¦¬ ë¦¬ìŠ¤íŠ¸)
íšŒì› ì •ë³´ê°€ ë‹´ê¸´ **ë”•ì…”ë„ˆë¦¬ ë¦¬ìŠ¤íŠ¸**ê°€ ìˆìŠµë‹ˆë‹¤. ì—¬ê¸°ì„œ **ì„±ì¸(20ì‚´ ì´ìƒ)**ë§Œ ê³¨ë¼ë‚´ëŠ” ê¸°ëŠ¥ì„ ë§Œë“œì„¸ìš”.

**ë°ì´í„°:**
`users = [{'name': 'Kim', 'age': 15}, {'name': 'Lee', 'age': 25}, {'name': 'Park', 'age': 30}]`

**ë¯¸ì…˜:**
1. `users` ë¦¬ìŠ¤íŠ¸ë¥¼ `for`ë¬¸ìœ¼ë¡œ ìˆœíšŒí•˜ì„¸ìš”.
2. `if`ë¬¸ì„ ì‚¬ìš©í•´ ë‚˜ì´(`age`)ê°€ 20 ì´ìƒì¸ ì‚¬ëŒì˜ **ì´ë¦„(`name`)**ë§Œ ë”°ë¡œ ë¦¬ìŠ¤íŠ¸(`adults`)ì— ë‹´ìœ¼ì„¸ìš”.
3. `adults` ë¦¬ìŠ¤íŠ¸ë¥¼ ì¶œë ¥í•˜ì„¸ìš”. -> `['Lee', 'Park']`

```python
# [Advanced 2 ì‘ì„± ì…€]
users = [
    {'name': 'Kim', 'age': 15}, 
    {'name': 'Lee', 'age': 25}, 
    {'name': 'Park', 'age': 30}
]

adults = []

for ages in users:
    age_val = ages['age']
    if age_val >= 20:
        adults.append(ages['name'])

print(adults)

all_ages = [user['name'] for user in users if user['age'] >= 20]
print(all_ages)
```
