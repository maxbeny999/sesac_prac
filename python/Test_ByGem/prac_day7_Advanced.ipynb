### 문제 1: 은행 계좌 관리 시스템 (Bank Account)
**목표:** 입금, 출금, 이체 기능과 예외 처리(잔액 부족 등)를 포함한 완벽한 클래스를 구현하세요.

**조건:**
1. `BankAccount` 클래스를 정의합니다. (속성: `owner`, `balance`)
2. `deposit(amount)`: 입금액이 0보다 작으면 "잘못된 금액" 출력, 아니면 잔액 증가.
3. `withdraw(amount)`: 잔액보다 출금액이 크면 "잔액 부족" 출력, 아니면 잔액 감소.
4. `transfer(target, amount)`: 
   - **핵심:** `target`은 **다른 BankAccount 객체**입니다.
   - 내 잔액이 충분하면 내 돈을 줄이고(`withdraw`), 상대방 계좌에 돈을 넣으세요(`deposit`).
   - 성공 시 True, 실패(잔액부족 등) 시 False 반환.

```python
# [문제 1 작성 셀]

# 테스트 코드 예시
# acc1 = BankAccount("Alice", 5000)
# acc2 = BankAccount("Bob", 1000)
# acc1.transfer(acc2, 2000) 
# print(acc1.balance) # 3000
# print(acc2.balance) # 3000
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        if amount < 0:
            print( "잘못된 금액" )
        else:
            self.balance += amount
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액부족")
        else:
            self.balance -= amount
        
    def transfer(self, target, amount):
        if self.balance >= amount:
            self.balance -= amount
            target.deposit(amount)
            return True
        else:
            return False


```

### 문제 2: 도서관 책 관리 (List of Objects)
**목표:** `Book` 클래스와 `Library` 클래스를 정의하고, `Library`가 여러 `Book` 객체를 관리하도록 합니다.

**조건:**
1. `Book` 클래스:
   - 속성: `title`, `status` (기본값 "available")
   - 메서드: `borrow()` (상태를 "borrowed"로 변경), `return_book()` (상태를 "available"로 변경)

2. `Library` 클래스:
   - 속성: `books` (Book 객체들을 담을 빈 리스트)
   - `add_book(book)`: Book 객체를 리스트에 추가.
   - `borrow_book(title)`: 
     - 제목(`title`)으로 책을 찾고, 책이 있고 "available"이면 대출 처리(Book의 `borrow` 호출).
     - 이미 대출 중이면 "대출 중입니다" 출력, 책이 없으면 "책이 없습니다" 출력.

**힌트:** `Library`에서 책을 찾을 때 `for book in self.books:`를 사용하세요.

```python
# [문제 2 작성 셀]
class Book:
    def __init__(self, title, status):
        self.title = title
        self.status = status

    def borrow(self):
        self.status = "borrowed"

    def return_book(self):
        self.status = "available"

class Library:
    def __init__(self):
        self.books = []
    
    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and book.status == "available":
                book.borrow()
                print(f"{title} 대출 완료")
                return
        print("책이 없거나 대출 중입니다.")
```
