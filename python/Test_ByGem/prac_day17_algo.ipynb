## [Set 1] 문자열 다루기 (Bronze Level)
### Q1-1. (Warm-up) 문자열 뒤집기
주어진 문자열 `s`를 뒤집어서 반환하는 함수를 작성하세요.

```python
def reverse_string(s: str) -> str:
    reverse_s = s[:: -1]
    return reverse_s

# 테스트
print("Q1-1 결과:", reverse_string("fastapi")) # 예상 결과: "ipatsaf"
```

### Q1-2. (Main) 팰린드롬인지 확인하기
알파벳 소문자로만 이루어진 단어가 주어집니다. 이 단어가 팰린드롬(앞으로 읽을 때와 거꾸로 읽을 때 똑같은 단어)이면 1, 아니면 0을 반환하는 함수를 작성하세요. (백준 10988번 스타일)

```python
def is_palindrome(word: str) -> int:
    if word == word[:: -1]:
        return 1
    else:
        return 0

# 테스트
print("Q1-2 결과 (level):", is_palindrome("level")) # 예상 결과: 1
print("Q1-2 결과 (baekjoon):", is_palindrome("baekjoon")) # 예상 결과: 0
```

## [Set 2] 배열과 수학 (Bronze Level)
### Q2-1. (Warm-up) 최댓값과 인덱스 찾기
주어진 리스트 `nums`에서 최댓값과 그 최댓값이 몇 번째(0-index 기준)에 있는지 튜플로 반환하는 함수를 작성하세요.

```python
def find_max_and_index(nums: list) -> tuple:
        current_max = nums[0]
        max_index = 0
        for i, n in enumerate(nums):
            if n > current_max:
                current_max = n
                max_index = i
        return max_index, current_max
        

# 테스트
print("Q2-1 결과:", find_max_and_index([3, 29, 38, 12, 57, 74, 40, 85, 61])) # 예상 결과: (85, 7)
```

### Q2-2. (Main) 평균 조작하기
기말고사 점수 리스트가 주어집니다. 가장 높은 점수를 `M`이라고 할 때, 모든 점수를 `점수/M*100`으로 고쳤습니다. 이렇게 고친 새로운 점수들의 평균을 구하는 함수를 작성하세요. (백준 1546번 스타일)

```python
# def calculate_new_average(scores: list) -> float:
#         current_max = scores[0] 
#         new_average = []
#         for n in scores:
#             if n > current_max:
#                 current_max = n
#         M = current_max
#         for n in scores:
#             new_average.append(n / M*100)
#         answer = sum(new_average) / len(new_average)
#         return answer

def calculate_new_average(scores: list) -> float:
        M = max(scores)
        new_scores = [n / M*100 for n in scores]
        answer = sum(new_scores) / len(new_scores)
        return answer


# 테스트
print("Q2-2 결과 1:", calculate_new_average([40, 80, 60])) # 예상 결과: 75.0
print("Q2-2 결과 2:", calculate_new_average([10, 20, 30])) # 예상 결과: 66.66666666666667
```

### Q3-1. (Warm-up) 빈도수 딕셔너리 만들기
숫자들이 담긴 리스트 `cards`가 주어집니다. 각 숫자가 몇 번 등장하는지 세어서 딕셔너리로 반환하는 함수를 작성하세요.

```python
from collections import Counter
# def count_frequencies(cards: list) -> dict:
#     emp = {}
#     for nums in cards:
#         if nums not in emp:
#             emp[nums] = 1
#         else:  
#             emp[nums] +=1
#     return emp

# def count_frequencies(cards: list) -> dict:
#     emp ={}
#     for num in cards:
#         emp[num] = emp.get(num, 0) + 1
#     return emp

def count_frequencies(cards: list) -> dict:
    emp = dict(Counter(cards))
    return emp

# 테스트
print("Q3-1 결과:", count_frequencies([10, 9, -5, 2, 3, 4, 5, -5])) 
# 예상 결과: {10: 1, 9: 1, -5: 2, 2: 1, 3: 1, 4: 1, 5: 1}
```

### Q3-2. (Main) 숫자 카드 2
상근이가 가지고 있는 숫자 카드의 리스트 `cards`와, 몇 개인지 구해야 할 숫자들의 리스트 `targets`가 주어집니다. `targets`의 각 숫자가 상근이의 카드에 몇 개 있는지 순서대로 리스트에 담아 반환하세요. (백준 10816번 스타일)\n\n**제한사항:** 이중 for문을 사용하면 시간 초과가 발생할 수 있습니다. Q3-1에서 연습한 딕셔너리(Hash) 개념을 반드시 활용하세요.

```python

def count_target_cards(cards: list, targets: list) -> list:

    card = dict(Counter(cards))   
    return [card.get(nums, 0) for nums in targets]
# 테스트
cards_list = [6, 3, 2, 10, 10, 10, -10, -10, 7, 3]
targets_list = [10, 9, -5, 2, 3, 4, 5, -10]
print("Q3-2 결과:", count_target_cards(cards_list, targets_list)) 
# 예상 결과: [3, 0, 0, 1, 2, 0, 0, 2]
```
