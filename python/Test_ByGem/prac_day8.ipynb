### [Basic 3] ë¬¸ìì—´ íŒŒì‹±ê³¼ ë³€í™˜
ë¬¸ìì—´ `data = "apple:1000,banana:2000,cherry:3000"`ì´ ì£¼ì–´ì§‘ë‹ˆë‹¤.
ì´ ë¬¸ìì—´ì„ íŒŒì‹±í•˜ì—¬ ì•„ë˜ì™€ ê°™ì€ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜í•˜ì—¬ ì¶œë ¥í•˜ì„¸ìš”.

**ì¶œë ¥ ì˜ˆì‹œ:** `{'apple': 1000, 'banana': 2000, 'cherry': 3000}`
*(íŒíŠ¸: `split`ì„ ë‘ ë²ˆ ì‚¬ìš©í•˜ê³ , ê°€ê²©ì€ ì •ìˆ˜(`int`)ë¡œ ë³€í™˜í•´ì•¼ í•©ë‹ˆë‹¤.)*

````python
data = "apple:1000,banana:2000,cherry:3000"
# ì‘ì„±
word = data.split(",")
answer = {}
for words in word:
    key, value = words.split(":")
    value = int(value)
    answer[key] = value
print(answer)
````

### [Basic 4] ì¤‘ì²© ë¦¬ìŠ¤íŠ¸ í‰íƒ„í™” (Flatten)

### [Basic 7] í•¨ìˆ˜: ê°€ë³€ ì¸ì í™œìš©
ì…ë ¥ë°›ì€ ìˆ«ìë“¤ì˜ **í‰ê· **ì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ `calc_avg(*args)`ë¥¼ ì‘ì„±í•˜ì„¸ìš”.
ì¸ìì˜ ê°œìˆ˜ëŠ” ëª‡ ê°œê°€ ë“¤ì–´ì˜¬ì§€ ëª¨ë¦…ë‹ˆë‹¤.

```python
print(calc_avg(10, 20, 30))  # 20.0 ì¶œë ¥
```

````python
# ì‘ì„±
def calc_avg(*args):
    answer = sum(args) / len(args)
    return answer

print(calc_avg(10, 20, 30))  # 20.0 ì¶œë ¥
print(calc_avg(4, 6))   
````

### [Basic 8] ë”•ì…”ë„ˆë¦¬ ê°’ ê°±ì‹  (ë¹ˆë„ìˆ˜ ì„¸ê¸°)
ë¬¸ìì—´ ë¦¬ìŠ¤íŠ¸ê°€ ì£¼ì–´ì¡Œì„ ë•Œ, ê° ë‹¨ì–´ê°€ ëª‡ ë²ˆ ë“±ì¥í–ˆëŠ”ì§€ ë”•ì…”ë„ˆë¦¬ì— ì €ì¥í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
`words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']`

**ì¶œë ¥ ì˜ˆì‹œ:** `{'apple': 3, 'banana': 2, 'orange': 1}`

````python
words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
# ì‘ì„±
answer = {}
for fruits in words :
    if fruits not in answer:
        answer[fruits] = 1
    else:
        answer[fruits] += 1
print(answer)

words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
answer = {}

for fruit in words:
    # "fruit í‚¤ê°€ ìˆìœ¼ë©´ ê·¸ ê°’ì„ ê°€ì ¸ì˜¤ê³ , ì—†ìœ¼ë©´ 0ì„ ê°€ì ¸ì™€ì„œ +1 í•´ë¼"
    answer[fruit] = answer.get(fruit, 0) + 1

print(answer)
# ì¶œë ¥: {'apple': 3, 'banana': 2, 'orange': 1}
````
