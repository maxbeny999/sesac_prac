## ğŸ”´ Advanced: í´ë˜ìŠ¤ì™€ ëª¨ë“ˆì˜ ë§Œë‚¨

### [Advanced 1] ë¡œë˜ ê¸°ê³„ í´ë˜ìŠ¤
ì–´ì œ ë°°ìš´ í´ë˜ìŠ¤ì™€ ì˜¤ëŠ˜ ë°°ìš´ `random`ì„ í•©ì³ë´…ì‹œë‹¤.

**ë¯¸ì…˜:**
`LottoMachine` í´ë˜ìŠ¤ë¥¼ ë§Œë“œì„¸ìš”.
1. **ì†ì„±(`__init__`):** ì•„ë¬´ëŸ° ì†ì„±ë„ ë°›ì§€ ì•ŠìŠµë‹ˆë‹¤. ê·¸ëƒ¥ `pass` í•˜ê±°ë‚˜ ë¹„ì›Œë‘ì„¸ìš”.
2. **í–‰ë™(`pick_numbers`):**
    * 1ë¶€í„° 45 ì‚¬ì´ì˜ ìˆ«ì ì¤‘ **ì¤‘ë³µ ì—†ì´ 6ê°œ**ë¥¼ ë½‘ì•„ì•¼ í•©ë‹ˆë‹¤.
    * **íŒíŠ¸:** `random.sample(ë²”ìœ„, ê°œìˆ˜)` í•¨ìˆ˜ë¥¼ ì°¾ì•„ë³´ê±°ë‚˜, `set`ì„ ì“°ê±°ë‚˜, `while`ë¬¸ì„ ì¨ì„œ ì¤‘ë³µì„ í”¼í•´ë³´ì„¸ìš”.
    * ë½‘ì€ ìˆ«ì ë¦¬ìŠ¤íŠ¸ë¥¼ `return`(ë°˜í™˜) í•˜ì„¸ìš”.
    * (ì •ë ¬í•´ì„œ ì£¼ë©´ ë” ì¢‹ìŠµë‹ˆë‹¤!)

```python
# [Advanced 1 ì‘ì„± ì…€]
import random

class LottoMachine:
    def pick_numbers(self):
        lotto = random.sample(range(1, 46), 6)
        lotto.sort()
        return lotto

# í…ŒìŠ¤íŠ¸
lotto = LottoMachine()
print(lotto.pick_numbers())
```

### [Advanced 2] í¸ì˜ì  ì‚¬ì¥ë‹˜ (ì˜ˆì™¸ ì²˜ë¦¬ ë³µìŠµ)
`try-except`ë¥¼ í™œìš©í•´ ë¬¼ê±´ì„ ê³„ì‚°í•˜ëŠ” í´ë˜ìŠ¤ë¥¼ ë§Œë“­ë‹ˆë‹¤.

**ë¯¸ì…˜:**
`Counter` í´ë˜ìŠ¤ë¥¼ ë§Œë“œì„¸ìš”.
1. `price`(ë¬¼ê±´ ê°€ê²©)ì„ ì…ë ¥ë°›ì•„ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.
2. `calculate(money)` ë©”ì„œë“œ:
    * ì†ë‹˜ì´ ë‚¸ ëˆ(`money`)ì„ ë°›ìŠµë‹ˆë‹¤.
    * ê±°ìŠ¤ë¦„ëˆ(`money - price`)ì„ ê³„ì‚°í•´ ì¶œë ¥í•©ë‹ˆë‹¤.
    * **ì˜ˆì™¸ 1:** ë§Œì•½ ëˆì´ ë¶€ì¡±í•˜ë©´ "ëˆì´ ë¶€ì¡±í•©ë‹ˆë‹¤." ì¶œë ¥ (ifë¬¸ í™œìš©)
    * **ì˜ˆì™¸ 2:** ë§Œì•½ `money`ì— ìˆ«ìê°€ ì•„ë‹ˆë¼ ë¬¸ì("ì˜¤ì²œì›")ê°€ ë“¤ì–´ì˜¤ë©´, ì—ëŸ¬ê°€ ë‚˜ì„œ ë©ˆì¶”ì§€ ë§ê³  "ìˆ«ìë§Œ ë‚´ì„¸ìš”."ë¼ê³  ì¶œë ¥ (`try-except` í™œìš©)

```python
# [Advanced 2 ì‘ì„± ì…€]
class Counter:
    def __init__(self, price):
        self.price = price

    def calculate(self, money):
        self.money = money
        
        try:
            change = self.money - self.price
            if self.money < self.price :
                print("ëˆì´ ë¶€ì¡±í•©ë‹ˆë‹¤")
            else:
                print(change)

        except:
                print("ìˆ«ìë§Œ ë‚´ì„¸ìš”")

# í…ŒìŠ¤íŠ¸
c = Counter(3000)
c.calculate(5000)   # ì •ìƒ: 2000ì› ê±°ìŠ¬ëŸ¬ì¤Œ
c.calculate(1000)   # ë¶€ì¡±: ëˆ ë¶€ì¡± ë©”ì‹œì§€
c.calculate("ë§Œì›") # ì—ëŸ¬: ì˜ˆì™¸ ì²˜ë¦¬ ë©”ì‹œì§€
```
