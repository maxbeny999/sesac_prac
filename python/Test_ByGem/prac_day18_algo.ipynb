### 🏆 Main 1: 베스트셀러 (백준 1302번 스타일 / 실버 4)
오늘 하루 동안 팔린 책의 제목들이 리스트로 주어집니다. **가장 많이 팔린 책의 제목**을 출력하세요.
단, 가장 많이 팔린 책이 여러 개일 경우 **알파벳 순으로 가장 앞서는 제목**을 출력해야 합니다.

---
💡 **[Advanced Tip]**
딕셔너리로 빈도수를 구한 뒤, `max()` 함수의 `key` 속성을 활용하거나, 리스트를 특정 조건(판매량 내림차순, 제목 오름차순)으로 정렬(`sort`)하면 파이썬스럽게 O(N log N)으로 해결할 수 있습니다.

```python
from collections import Counter
books = ["top", "top", "top", "top", "kim", "kim", "kim", "kim", "abc", "abc"]

# 여기에 코드를 작성하세요.
book_counter = Counter(books)
sorted_books = sorted(book_counter.items(), key= lambda x:(-x[1], x[0]))
print(sorted_books[0][0])
```

### 🏆 Main 2: 장부 정리 (백준 10773번 '제로' 스타일 / 실버 4)
가계부에 숫자를 적고 있습니다. 숫자를 잘못 적었을 때는 `0`을 외치며 **가장 최근에 쓴 숫자를 지웁니다**.
모든 입력을 마친 후, 가계부에 적힌 수의 **총합**을 구하세요.
예: `[3, 0, 4, 0, 5, 6, 0]` -> `[5]` -> 총합 5

---
💡 **[Advanced Tip]**
'가장 최근에 들어온 데이터를 먼저 처리한다'는 전형적인 LIFO(Last In First Out) 구조입니다. 반복문과 Stack(리스트)을 조합하여, 순회가 끝난 후 최종 리스트에 `sum()`을 씌워보세요.

```python
calls = [100000, 20000, 0, 0, 50000, 10000, 0, 30000]

# 여기에 코드를 작성하세요.
def counter(calls): 
    stack =[]
    for char in calls:
        if char != 0:
            stack.append(char)
        else:
            stack.pop()
    return sum(stack)
print(counter(calls))
```
