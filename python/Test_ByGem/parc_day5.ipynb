### 문제 4: 투표 결과 세기 (List -> Dict)
**목표:**
1. `votes` 리스트에 투표 결과가 들어있습니다.
2. `for`문을 돌면서 누구에게 몇 표가 나왔는지 `result` 딕셔너리에 정리하세요.
   - 예: `{'아이유': 2, '유재석': 1}`

**힌트:**
- 딕셔너리에 이름이 **없으면** `1` 대입, **있으면** `+1` 증가.

```python
# [문제 4 작성 셀]
votes = ["아이유", "유재석", "박명수", "아이유", "유재석", "아이유"]
result = {}

for vote in votes:
    if vote not in result:
        result[vote] = 1
    else:
        result[vote] +=1
        

print(result)
```

### 문제 7: 딕셔너리 뒤집기 (For + Dict)
**목표:**
1. `word_dict`는 `한글:영어` 로 되어 있습니다.
2. 이를 뒤집어서 `영어:한글` 로 된 새로운 딕셔너리 `reversed_dict`를 만드세요.
   - 예: `{'사과':'apple'}` -> `{'apple':'사과'}`

```python
# [문제 7 작성 셀]
word_dict = {"사과": "apple", "바나나": "banana", "포도": "grape"}
reversed_dict = {}

for korean, english in word_dict.items():
    # korean, english = english, korean
    reversed_dict[english] = korean


print(reversed_dict)

reversed_dict = {english: korean for korean, english in word_dict.items()}

print(reversed_dict)
```

### 문제 8: 특정 값 삭제하기 (While + List)
**목표:**
1. `nums` 리스트에 숫자 **3**이 여러 개 들어있습니다.
2. `while` 문을 사용하여 리스트 안에 **3이 있는 동안(`in`)** 계속해서 3을 삭제(`remove`)하세요.
   - 힌트: `while 3 in nums:`

```python
# [문제 8 작성 셀]
nums = [1, 3, 2, 3, 4, 3, 5]

while 3 in nums:
    nums.remove(3)


print(nums) # 예상: [1, 2, 4, 5]
```

### 문제 9: 장바구니 총액 구하기 (List of Dicts)
**목표:**
1. `cart`는 딕셔너리들이 들어있는 **리스트**입니다.
2. `for`문을 돌면서 모든 물건의 가격(`price`)을 합산하여 `total_price`를 구하세요.

```python
# [문제 9 작성 셀]
cart = [
    {"name": "노트북", "price": 1000},
    {"name": "마우스", "price": 50},
    {"name": "키보드", "price": 150}
]
total_price = 0

for stuff in cart:
    total_price += stuff['price']


print(f"총 결제 금액: {total_price}")
```

### 문제 10: 영어 단어 퀴즈 (While + Dict)
**목표:**
1. `quiz` 딕셔너리에는 `영어단어:뜻` 이 들어있습니다.
2. 반복문으로 사용자에게 뜻(`value`)을 보여주고, 영어 단어(`key`)를 입력받으세요.
3. 정답이면 "정답!", 틀리면 "땡!"을 출력하세요.
4. 모든 문제를 다 풀면 반복을 종료하세요.


```python
# [문제 10 작성 셀]
quiz = {
    "apple": "사과",
    "banana": "바나나"
}
for eng, meaning in quiz.items():
    print(f"{eng}의 뜻은?")
    
    user_answer = input()

    if user_answer == meaning :
        print("정답!")
    else:
        print(f"땡! 정답은{meaning}")
```
