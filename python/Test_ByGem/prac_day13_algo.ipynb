### ğŸ”¥ Warm-up 11: ê´„í˜¸ ì§ ë§ì¶”ê¸° (Logic)
ì˜¬ë°”ë¥¸ ê´„í˜¸ ë¬¸ìì—´ì¸ì§€ íŒë‹¨í•˜ëŠ” ë¡œì§ì„ ì—°ìŠµí•´ë´…ë‹ˆë‹¤.
1. `(`ë¥¼ ë§Œë‚˜ë©´ ìŠ¤íƒì— ë„£ìŠµë‹ˆë‹¤.
2. `)`ë¥¼ ë§Œë‚˜ë©´ ìŠ¤íƒì—ì„œ í•˜ë‚˜ êº¼ëƒ…ë‹ˆë‹¤.
3. ë§Œì•½ êº¼ë‚¼ ê²Œ ì—†ëŠ”ë° `)`ê°€ ë‚˜ì˜¤ë©´? -> **"ì˜ëª»ëœ ë¬¸ìì—´" (False)**
4. ë‹¤ ëë‚¬ëŠ”ë° ìŠ¤íƒì— ë­”ê°€ ë‚¨ì•„ìˆìœ¼ë©´? -> **"ì˜ëª»ëœ ë¬¸ìì—´" (False)**
5. ìŠ¤íƒì´ í…… ë¹„ì–´ìˆìœ¼ë©´? -> **"ì˜¬ë°”ë¥¸ ë¬¸ìì—´" (True)**

```python
# ì˜ˆì‹œ: "(())" -> True
# ì˜ˆì‹œ: "(()"  -> False (ë‚¨ì•„ìˆìŒ)
# ì˜ˆì‹œ: ")("   -> False (êº¼ë‚¼ ê²Œ ì—†ìŒ)

s = "(())"
stack = []
is_valid = True

for char in s:
    if char == '(':
        stack.append(char)
    else: # char == ')'
        if not stack: # ìŠ¤íƒì´ ë¹„ì–´ìˆìœ¼ë©´
            is_valid = False
            break
        stack.pop()

if stack: # ë‹¤ ëŒì•˜ëŠ”ë° ë‚¨ì•„ìˆìœ¼ë©´
    is_valid = False

print(is_valid)
```

### ë¬¸ì œ 12: ê´„í˜¸ ê²€ì‚¬ (Silver / ë°±ì¤€ 9012)
Warm-up 11ì—ì„œ ë°°ìš´ ë¡œì§ì„ í•¨ìˆ˜ë¡œ ë§Œë“œì„¸ìš”.
ì£¼ì–´ì§„ ë¬¸ìì—´ì´ ì˜¬ë°”ë¥¸ ê´„í˜¸ ë¬¸ìì—´(VPS)ì´ë©´ `"YES"`, ì•„ë‹ˆë©´ `"NO"`ë¥¼ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ `check_vps(s)`ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

*í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤:*
1. `"(())())` -> NO
2. `"(((()())())"` -> NO
3. `"(()())((()))"` -> YES

```python
def check_vps(s):
    stack = []
    for char in s:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if not stack: # ë¹„ì–´ìˆëŠ”ë° ë‹«ëŠ” ê´„í˜¸ê°€ ë‚˜ì˜¤ë©´
                return "NO"
            stack.pop()
    
    # ë°˜ë³µë¬¸ì´ ëë‚œ ë’¤ ìŠ¤íƒ í™•ì¸
    if stack:
        return "NO"
    else:
        return "YES"

# í…ŒìŠ¤íŠ¸
print(check_vps("(())"))
print(check_vps("(())"))
print(check_vps("(())"))
```
