### ë¬¸ì œ 10: ì˜¬ë°”ë¥¸ ê´„í˜¸ ê²€ì‚¬ (Stack)
ìŠ¤íƒ(Stack) ìë£Œêµ¬ì¡°ì˜ ê°€ì¥ ê³ ì „ì ì¸ ë¬¸ì œì…ë‹ˆë‹¤.
ì£¼ì–´ì§„ ë¬¸ìì—´ `s`ê°€ ì˜¬ë°”ë¥¸ ê´„í˜¸ ìŒ `()`ì„ ì´ë£¨ê³  ìˆëŠ”ì§€ íŒë‹¨í•˜ì„¸ìš”.

**ë¡œì§ íŒíŠ¸:**
1. `(`ë¥¼ ë§Œë‚˜ë©´ ë¦¬ìŠ¤íŠ¸(ìŠ¤íƒ)ì— ë„£ìŠµë‹ˆë‹¤ (`append`).
2. `)`ë¥¼ ë§Œë‚˜ë©´ ë¦¬ìŠ¤íŠ¸ì—ì„œ í•˜ë‚˜ë¥¼ ëºë‹ˆë‹¤ (`pop`).
   - ë‹¨, ëº„ ê²Œ ì—†ìœ¼ë©´(ë¹„ì–´ìˆìœ¼ë©´) `False`ì…ë‹ˆë‹¤.
3. ëª¨ë“  ê³¼ì •ì´ ëë‚¬ëŠ”ë° ë¦¬ìŠ¤íŠ¸ì— ë­”ê°€ ë‚¨ì•„ìˆìœ¼ë©´ `False`ì…ë‹ˆë‹¤.
4. ê¹”ë”í•˜ê²Œ ë¹„ì–´ìˆìœ¼ë©´ `True`ì…ë‹ˆë‹¤.

```python
def is_valid_parenthesis(s):
    stack = []
    for element in s:
        if element == '(':
            stack.append(element)
        elif element == ')' and len(stack) != 0:
            stack.pop()
        elif len(stack) == 0:
            return False
    
    return len(stack) == 0 # ìˆ˜ì • í•„ìš”í•˜ë©´ ìˆ˜ì •í•˜ì„¸ìš”

# í…ŒìŠ¤íŠ¸ ì½”ë“œ
print(is_valid_parenthesis("(())()"))  # True
print(is_valid_parenthesis("(()"))     # False (ì•ˆ ë‹«í˜)
print(is_valid_parenthesis(")("))      # False (ìˆœì„œ í‹€ë¦¼)
print(is_valid_parenthesis("((()))"))  # True
```
