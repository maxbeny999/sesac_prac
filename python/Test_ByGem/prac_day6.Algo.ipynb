### ⚔️ 문제 1: 카드2 (Queue 활용)
**문제 설명:**
N장의 카드가 있습니다. 카드는 차례대로 1부터 N까지의 번호가 붙어 있으며, 1번 카드가 제일 위에, N번 카드가 제일 아래인 상태로 순서대로 쌓여있습니다.

다음 동작을 카드가 **한 장 남을 때까지** 반복합니다.
1. 제일 위에 있는 카드를 바닥에 버립니다. (Pop)
2. 그 다음, 제일 위에 있는 카드를 제일 아래로 옮깁니다. (Move to Back)

N이 주어졌을 때, 제일 마지막에 남게 되는 카드를 구하세요.

**입력 예시:**
`6` (1, 2, 3, 4, 5, 6)

**과정 시뮬레이션:**
- 1 버림 -> (2, 3, 4, 5, 6) -> 2를 뒤로 -> (3, 4, 5, 6, 2)
- 3 버림 -> (4, 5, 6, 2) -> 4를 뒤로 -> (5, 6, 2, 4)
- ... (반복)

**힌트:**
- 일반 `list`의 `pop(0)`을 쓰면 속도가 느립니다.
- 파이썬의 사기템 `from collections import deque`를 사용하세요.
- `deque`에는 `popleft()`(왼쪽 빼기)와 `append()`(오른쪽 넣기)가 있습니다.

```python
from collections import deque

def solve_card2(n):
    # 1부터 N까지 들어있는 deque를 만드세요
    # queue = deque([1, 2, 3 ...]) 또는 range 활용
    queue = deque(range(1,n+1))
    while len(queue) > 1:
        queue.popleft()
        queue.append(queue.popleft())
    return queue[0]

# 테스트
print(solve_card2(6)) # 예상 결과: 4
```

### ⚔️ 문제 2: 단어 공부 (Dict + Logic)
**문제 설명:**
알파벳 대소문자로 된 단어가 주어지면, 이 단어에서 **가장 많이 사용된 알파벳**이 무엇인지 알아내는 프로그램을 작성하세요.
단, 대문자와 소문자를 구분하지 않으며(a와 A는 같은 글자 취급), 결과는 **대문자**로 출력해야 합니다.

**조건:**
1. 가장 많이 사용된 알파벳이 여러 개 존재하는 경우에는 `?`를 반환하세요.

**입력 예시:**
1. `Mississipi` -> `?` (i와 s가 똑같이 4번씩 나옴)
2. `zZa` -> `Z` (Z가 2번, a가 1번)

**힌트:**
- 먼저 모든 문자를 대문자로 바꾸세요. (`.upper()`)
- 딕셔너리에 개수를 세서 넣으세요.
- 개수(Value) 중 최대값을 찾고, 그 최대값을 가진 알파벳(Key)이 몇 개인지 확인하세요.

```python
def study_word(word):
    # 1. 전부 대문자로 변환 (a와 A를 똑같이 취급하기 위해)
    word = word.upper() 
    
    # 2. 딕셔너리에 개수 카운팅 (Day 5에서 배운 로직!)
    count_dict = {}
    for char in word:
        if char not in count_dict:
            count_dict[char] = 0
        count_dict[char] += 1
    
    # 3. 최대값 찾기 및 공동 1등 처리
    max_count = 0  # 현재까지 확인한 가장 많이 나온 횟수
    result = "?"   # 최종 정답 (알파벳 혹은 ?)
    
    # 딕셔너리의 키(char)와 값(count)을 하나씩 꺼내서 확인
    for char, count in count_dict.items():
        # (1) 신기록 갱신! (단독 1등 후보)
        if count > max_count:
            max_count = count
            result = char
        # (2) 현재 1등과 횟수가 같음! (공동 1등)
        elif count == max_count:
            result = "?"
            
    return result

# 테스트
print(study_word("Mississipi")) # 결과: ?
print(study_word("zZa"))        # 결과: Z
```
