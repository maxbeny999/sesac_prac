### ë¬¸ì œ 1: í•œ ì¤„ë¡œ ë¦¬ìŠ¤íŠ¸ ë³€í™˜ (List Comprehension)
ë‹¤ìŒ `prices` ë¦¬ìŠ¤íŠ¸ì— ìžˆëŠ” ê°€ê²©ë“¤ì— **ë¶€ê°€ì„¸ 10%ë¥¼ ë”í•œ ê°€ê²©**ì„ ê³„ì‚°í•˜ì—¬ ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ `final_prices`ë¥¼ ë§Œë“œì„¸ìš”.
ë‹¨, `for` ë£¨í”„ë¥¼ ê¸¸ê²Œ ì“°ì§€ ë§ê³  **ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜**ì„ ì‚¬ìš©í•´ í•œ ì¤„ë¡œ ìž‘ì„±í•˜ì„¸ìš”.
(ì†Œìˆ˜ì ì€ ë‚¨ê²¨ë‘ì„¸ìš”)

```python
prices = [1000, 2500, 3000, 500]

# ì •ë‹µ ì½”ë“œë¥¼ í•œ ì¤„ë¡œ ìž‘ì„±í•˜ì„¸ìš”
# final_prices = [...]
final_prices = [x * 1.1 for x in prices ]

# ì¶œë ¥: [1100.0, 2750.0, 3300.0, 550.0]
print(final_prices)
```

### ë¬¸ì œ 2: ì¡°ê±´ì´ ìžˆëŠ” ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜
ë¦¬ìŠ¤íŠ¸ `words`ì—ì„œ ê¸¸ì´ê°€ **5ê¸€ìž ì´ìƒì¸ ë‹¨ì–´**ë§Œ ê³¨ë¼ì„œ, **ëª¨ë‘ ëŒ€ë¬¸ìž**ë¡œ ë³€í™˜í•œ ë¦¬ìŠ¤íŠ¸ `long_words`ë¥¼ ë§Œë“œì„¸ìš”.
ì´ë²ˆì—ë„ **ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜**ì„ ì‚¬ìš©í•˜ì„¸ìš”.

```python
words = ["apple", "cat", "banana", "dog", "elephant"]

# ížŒíŠ¸: [word.upper() for word in words if ...]
# long_words = 
long_words = [x.upper() for x in words if len(x) >= 5]
# ì¶œë ¥: ['APPLE', 'BANANA', 'ELEPHANT']
print(long_words)
```

### ë¬¸ì œ 3: ëžŒë‹¤(Lambda)ë¡œ ì •ë ¬í•˜ê¸°
íŠœí”Œë¡œ ì´ë£¨ì–´ì§„ ë¦¬ìŠ¤íŠ¸ `students`ê°€ ìžˆìŠµë‹ˆë‹¤. `(ì´ë¦„, ì ìˆ˜)` í˜•íƒœìž…ë‹ˆë‹¤.
**ì ìˆ˜(ë‘ ë²ˆì§¸ ìš”ì†Œ)ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬**í•˜ì„¸ìš”.
`sort()` ë˜ëŠ” `sorted()` í•¨ìˆ˜ì˜ `key` ì¸ìžì— **lambda í•¨ìˆ˜**ë¥¼ ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤.

```python
students = [("Alice", 85), ("Bob", 70), ("Charlie", 95)]

# ì ìˆ˜ ê¸°ì¤€ ì •ë ¬
# students.sort(key=lambda x: ...)
students.sort(key = lambda x: x[1])
# ì¶œë ¥: [('Bob', 70), ('Alice', 85), ('Charlie', 95)]
print(students)
```

### ë¬¸ì œ 4: ê°€ë³€ ì¸ìž (*args)
ìž…ë ¥ë°›ëŠ” ìˆ«ìžì˜ ê°œìˆ˜ê°€ ëª‡ ê°œë“  ìƒê´€ì—†ì´, ë“¤ì–´ì˜¨ ëª¨ë“  ìˆ«ìžë¥¼ ë”í•´ì„œ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ `sum_all`ì„ ë§Œë“œì„¸ìš”.
í•¨ìˆ˜ ì •ì˜ ì‹œ `*args`ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.

```python
def sum_all(*args):
    return sum(args)
# í…ŒìŠ¤íŠ¸
print(sum_all(1, 2, 3))       # 6
print(sum_all(10, 20, 30, 40)) # 100
```

### ë¬¸ì œ 5: í‚¤ì›Œë“œ ê°€ë³€ ì¸ìž (**kwargs)
ì‚¬ìš©ìž ì •ë³´ë¥¼ ìœ ì—°í•˜ê²Œ ë°›ì•„ì„œ ë”•ì…”ë„ˆë¦¬ë¡œ ë§Œë“¤ì–´ì£¼ëŠ” í•¨ìˆ˜ `make_profile`ì„ ë§Œë“œì„¸ìš”.
í•„ìˆ˜ ì¸ìžë¡œ `name`ì„ ë°›ê³ , ë‚˜ë¨¸ì§€ ì •ë³´ëŠ” `**kwargs`ë¡œ ë°›ì•„ì„œ í•©ì¹˜ì„¸ìš”.

*FastAPIì—ì„œ ì„¤ì •ì„ ë„˜ê¸¸ ë•Œ ì´ íŒ¨í„´ì„ ìžì£¼ ì”ë‹ˆë‹¤.*

```python
def make_profile(name, **kwargs):
    profile = {"name": name}
    # kwargsì— ë“¤ì–´ìžˆëŠ” ë‚´ìš©ë“¤ì„ profileì— ì—…ë°ì´íŠ¸í•˜ì„¸ìš”
    # ížŒíŠ¸: profile.update(...)
    
    return profile

# í…ŒìŠ¤íŠ¸
user1 = make_profile("Kim", age=25, city="Seoul")
print(user1) 
# ì¶œë ¥: {'name': 'Kim', 'age': 25, 'city': 'Seoul'}
```
