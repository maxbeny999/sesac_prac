## 문제 1: 활성 사용자 필터링
주어진 사용자 리스트에서 `is_active`가 `True`이면서 나이가 25세 이상인 사용자의 **이름만** 담긴 리스트를 만들어보세요.

---
💡 **[Advanced Tip]**
단순 `for`문과 빈 리스트를 추가하는 방식으로도 가능하지만, **리스트 컴프리헨션(List Comprehension)**을 사용하면 코드를 훨씬 파이썬스럽고(Pythonic) 간결하게 작성할 수 있습니다.

```python
users = [
    {"name": "Alice", "age": 24, "is_active": True},
    {"name": "Bob", "age": 27, "is_active": False},
    {"name": "Charlie", "age": 30, "is_active": True},
    {"name": "David", "age": 22, "is_active": True},
    {"name": "Eve", "age": 28, "is_active": True}
]

# 여기에 코드를 작성하세요.
answer = [names for names in users if names['age'] > 25 and names['is_active'] ]
print(answer)
```

## 문제 2: 재고 자산 계산
쇼핑몰 API에서 받아온 상품 재고 데이터입니다. 모든 상품의 `가격(price) * 수량(stock)`을 합산하여 총 재고 자산 가치를 구하세요.

---
💡 **[Advanced Tip]**
딕셔너리의 `.values()`를 순회하면서 계산할 수 있습니다. `sum()` 함수와 제너레이터 표현식을 결합하면 시간 복잡도 O(N)으로 한 줄에 우아하게 풀 수 있답니다.

```python
inventory = {
    "laptop": {"price": 1200000, "stock": 5},
    "mouse": {"price": 35000, "stock": 20},
    "keyboard": {"price": 89000, "stock": 10}
}

# 여기에 코드를 작성하세요.
total = sum(item['price'] * item['stock'] for item in inventory.values() )
print(total)
```

## 문제 3: 상태 코드별 그룹화
API 서버의 요청 로그 리스트가 주어집니다. 각 상태 코드(status)를 키(key)로 하고, 해당 상태 코드를 가진 로그의 ID 리스트를 값(value)으로 하는 딕셔너리를 만드세요.

---
💡 **[Advanced Tip]**
`if`문을 사용해 키가 존재하는지 매번 검사할 수도 있지만, `collections` 내장 모듈의 `defaultdict`를 활용하면 키 에러(KeyError) 걱정 없이 로직을 훨씬 단순화할 수 있습니다.

```python
from collections import defaultdict

logs = [
    {"id": 1, "status": 200}, {"id": 2, "status": 404},
    {"id": 3, "status": 200}, {"id": 4, "status": 500},
    {"id": 5, "status": 404}, {"id": 6, "status": 200}
]

# 여기에 코드를 작성하세요.
group_code = defaultdict(list)

for data in logs:
        group_code[data['status']].append(data['id'])
print(dict(group_code))
```

## 문제 4: 설정 값 병합 (Overriding)
앱의 기본 설정값(`default_settings`)과 사용자가 변경한 설정값(`user_settings`)이 있습니다. 최종 설정값을 담은 딕셔너리를 생성하세요. 단, 사용자가 변경한 값이 있다면 기본값을 덮어써야 합니다.

---
💡 **[Advanced Tip]**
`for`문으로 순회하며 값을 덮어써도 되지만, 딕셔너리의 `.update()` 메서드나 언패킹 연산자(`**`), 혹은 파이썬 3.9 이상에서 지원하는 병합 연산자(`|`)를 사용하면 손쉽게 해결할 수 있습니다.

```python
default_settings = {"theme": "light", "notifications": True, "language": "en"}
user_settings = {"theme": "dark", "language": "ko"}

# 여기에 코드를 작성하세요.
default_settings.update(user_settings)
print(default_settings)
```

## 문제 5: 중첩 JSON 데이터 안전하게 파싱하기
외부 API에서 아래와 같이 복잡하게 중첩된 데이터를 받았습니다. 이 데이터에서 사용자의 이메일 주소를 추출하세요. 만약 데이터 구조상 `profile`, `contact` 또는 `email` 키가 존재하지 않는다면 에러를 내지 않고 `None`을 반환해야 합니다.

---
💡 **[Advanced Tip]**
중첩 딕셔너리에서 `[key]` 방식으로 접근하면 구조가 바뀔 때 `KeyError`가 발생해 서버가 다운될 수 있습니다. 딕셔너리의 `.get()` 메서드를 체이닝하여 안전하게 접근하는 현업의 필수 습관을 들여보세요.

```python
api_response = {
    "status": "success",
    "data": {
        "user": {
            "id": 101,
            "profile": {
                "contact": {"email": "max@sesac.com", "phone": "010-1234-5678"}
            }
        }
    }
}

# 여기에 코드를 작성하세요.
print(api_response["data"]['user'].get('profile', {}).get('contact', {}).get('email'))
```

## 문제 6: 빈 시간대 찾기 (2D List)
회의실의 3일간 예약 상태가 2차원 리스트로 주어집니다. `0`은 예약 가능, `1`은 예약 완료를 뜻합니다.
예약 가능한 모든 시간대의 인덱스 `(일차, 시간대)`를 리스트 안의 튜플 형태로 반환하세요. (인덱스는 0부터 시작합니다.)
출력 예시: `[(0, 2), (0, 3), (1, 0), ...]`

---
💡 **[Advanced Tip]**
이중 `for`문으로 처리할 때, `enumerate()` 내장 함수를 활용하면 별도의 변수를 선언하여 카운팅하지 않고도 '요일 인덱스'와 '시간대 인덱스'를 직관적이고 깔끔하게 가져올 수 있습니다.

```python
schedule = [
    [1, 1, 0, 0, 1], # 1일차
    [0, 1, 1, 0, 0], # 2일차
    [1, 0, 0, 0, 1]  # 3일차
]

# 예약 가능한 시간대의 좌표를 담을 빈 장바구니를 준비합니다.
available_times = []

# [바깥쪽 루프] 전체 표에서 '하루치 일정(줄)'을 하나씩 꺼냅니다.
# enumerate() 덕분에 줄 번호(day_idx)와 그 줄의 데이터(day_schedule)를 동시에 받습니다.
for day_idx, day_schedule in enumerate(schedule):
    
    # [안쪽 루프] 꺼내온 하루치 일정 안에서 '각 시간대의 상태(칸)'를 하나씩 꺼냅니다.
    # 다시 한번 enumerate()를 써서 칸 번호(time_idx)와 상태(is_booked)를 동시에 받습니다.
    for time_idx, is_booked in enumerate(day_schedule):
        
        # 만약 그 시간대의 상태가 0 (예약 가능) 이라면?
        if is_booked == 0:
            # 줄 번호(일차)와 칸 번호(시간대)를 튜플 () 로 예쁘게 묶어서 장바구니에 담습니다.
            available_times.append((day_idx, time_idx))

print(available_times)
# 출력: [(0, 2), (0, 3), (1, 0), (1, 3), (1, 4), (2, 1), (2, 2), (2, 3)]
```
