 ## 문제 1: 구구단 중 '홀수단'만 출력하기 (이중 for문)

 **목표:**
 1. 2단부터 9단까지 구구단을 출력합니다.
 2. 단, **홀수 단(3, 5, 7, 9단)**만 출력해야 합니다.
 3. 추가 조건: 각 단의 결과값(곱한 값)이 **20 이하**인 경우만 출력하세요.

 **출력 예시:**
 === 3단 ===
 3 x 1 = 3
 ...
 3 x 6 = 18
 (21부터는 출력 X)
 === 5단 ===
 ...

```python
# [문제 1 작성 셀]
print("--- 구구단 홀수단 & 20 이하 출력 ---")
n = 1

for nums in range(2, 10):

    if nums % 2 != 0:
        for multiples in range(1, 10):
            multi_num = nums * multiples
            if multi_num < 20:
                print(f"{nums}x{multiples} = {multi_num} ")
            else:
                break
        
```


 ## 문제 2: 중복 제거하고 정렬하기 (List Logic)

 **목표:**
 `data_list`에는 중복된 숫자들이 섞여 있습니다.
 1. **중복을 제거**한 리스트를 만드세요. (set() 함수 사용 금지, for/if 사용)
 2. **오름차순(작은 수 -> 큰 수)으로 정렬**하세요. (sort() 함수 사용 금지, 직접 로직 구현)
 3. 최종 리스트를 출력하세요.

```python
# [문제 2 작성 셀]
data_list = [10, 50, 20, 10, 30, 50, 40, 20]

unique_list = []

for unique_num in data_list:
    if unique_num not in unique_list:
        unique_list.append(unique_num)
print(f"정렬전 {unique_list}")

n = len(unique_list)
for i in range(n):
    for j in range(0, n - 1):

        if unique_list[j] > unique_list[j+1]:
            unique_list[j], unique_list[ j +1] = unique_list[j+1], unique_list[j]
print(f"정렬후 {unique_list}")

```

 ## 문제 3: 자판기 거스름돈 계산 (Greedy Algorithm)

 **목표:**
 1. 손님이 낸 돈(`money`)에서 물건 값(`price`)을 뺍니다.
 2. 돈이 부족하면 "돈이 부족합니다"를 출력하세요.
 3. 돈이 남으면, 거스름돈을 줍니다. 단, **동전 개수를 최소한**으로 줄여야 합니다.
    (가능한 500원을 먼저 다 주고, 나머지를 100원으로 줍니다.)
 4. 최종적으로 500원짜리 몇 개, 100원짜리 몇 개인지 출력하세요.

```python
# [문제 3 작성 셀]
money = 3200
price = 1100

change = money - price

coin_5 = 500
coin_1 = 100

if money < price:
    print("돈이 부족합니다")
else:
    change5 = change // coin_5
    # left_change = change - change5 * coin_5
    left_change = change % coin_5
    change1 = left_change // coin_1
    print(change5, change1)
```

 ## 문제 4: 리스트 안의 딕셔너리 필터링 (Data Processing)

 **목표:**
 `students` 리스트에서 다음 조건에 맞는 학생의 **이름**만 추출하여 `passed_students` 리스트에 담고 출력하세요.

 **합격 조건:**
 1. 평균 점수 `(math + eng) / 2` 가 **80점 이상**이어야 함.
 2. **과락 제도:** 평균이 아무리 높아도, 수학(math) 또는 영어(eng) 중 하나라도 **60점 미만**이면 탈락.

```python
# [문제 4 작성 셀]
students = [
    {"name": "Alice", "math": 90, "eng": 85},
    {"name": "Bob", "math": 100, "eng": 50},
    {"name": "Charlie", "math": 80, "eng": 80},
    {"name": "David", "math": 40, "eng": 40},
    {"name": "Eve", "math": 95, "eng": 95}
]
passed_students = []
for high_graders in students:
    math_score = high_graders['math']
    eng_score = high_graders['eng']
    if (math_score + eng_score) / 2 >= 80 and math_score > 60 and eng_score > 60:
        passed_students.append(high_graders['name'])
print(passed_students)
```

 ## 문제 5: 다이아몬드 별 찍기 (Logic)

 **목표:**
 `level` 변수에 따라 피라미드 모양의 별을 출력하세요.
 `for`문과 문자열 연산을 사용하여 구현합니다.

 **예시 (level=3):**
   * (공백 2개, 별 1개)
  *** (공백 1개, 별 3개)
 ***** (공백 0개, 별 5개)

```python
# [문제 5 작성 셀]
level = 5


for i in range (level):
    spaces = " " * (level -1 -i)
    stars = "*" * ( 2 * i + 1 )
    print(spaces + stars)

# 힌트: 아래쪽 층수는 (level - 1) 층이어야 가운데가 안 겹치고 예쁩니다.
# range의 변수를 i 말고 j 같은 걸로 써보세요.

for n in range (level -1):
    spaces = " " *(n + 1)
    stars = "*" * (7 - (2 * n) )
    print( spaces + stars )
```

 ## 문제 6: 명예의 전당 (Top 3 Ranking)
 **난이도:** ⭐⭐⭐ (방금 배운 정렬 심화 복습)

 **목표:**
 게임 플레이어들의 점수 리스트(`scores`)가 있습니다.
 1. **중복된 점수를 제거**하세요. (한 사람이 기록한 중복 점수는 하나만 남김)
 2. 점수가 **높은 순서대로(내림차순)** 정렬하세요. (예: 100 -> 95 -> 80...)
 3. 정렬된 리스트에서 **상위 3개**의 점수만 뽑아서 `top3_scores`에 담으세요.

 **힌트:**
 - **정렬 로직:** 아까는 "앞사람이 더 크면" 바꿨죠? 이번엔 **"앞사람이 뒷사람보다 작으면"** 뒤로 보내야 큰 사람이 앞으로 오겠죠? (`<` 사용)
 - **리스트 자르기:** 슬라이싱(`[:3]`)을 쓰거나 `append`를 3번 하면 됩니다.

```python
# [문제 6 작성 셀]
scores = [50, 80, 100, 50, 95, 80, 70, 20, 100]

final_ranking = [] # 중복 제거 후 정렬된 전체 리스트를 담을 곳
top3_scores = []   # 최종 상위 3명을 담을 곳

# 1. 중복 제거 로직 (for, if, not in 사용)
# 여기에 코드를 작성하세요.
for high_graders in scores:
    if high_graders not in final_ranking:
        final_ranking.append(high_graders)


# 2. 내림차순 정렬 로직 (이중 for문, Bubble Sort 응용)
# 힌트: n = len(final_ranking) 으로 길이를 먼저 구하세요.
# 여기에 코드를 작성하세요.

n = len(final_ranking)
for i in range(n):
    for j in range(0, n - 1):
        if final_ranking[j] < final_ranking[j+1]:
            final_ranking[j], final_ranking[j+1] = final_ranking[j+1], final_ranking[j]
print(final_ranking)
# 3. 상위 3개 추출 로직
# 여기에 코드를 작성하세요.
top3_scores = final_ranking[0:3]


# 결과 확인
print(f"전체 랭킹(중복제거): {final_ranking}")
print(f"명예의 전당(Top 3): {top3_scores}")
```
