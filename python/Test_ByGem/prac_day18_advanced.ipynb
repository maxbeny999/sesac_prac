## 문제 1: Pydantic을 활용한 데이터 검증 (Data Validation)
FastAPI에서 클라이언트가 보낸 데이터를 안전하게 받기 위해 사용하는 `Pydantic` 모델을 만들어보세요.
상품을 뜻하는 `Product` 모델을 설계해야 합니다.

**[조건]**
1. `name`: 문자열 (필수)
2. `price`: 정수 (필수, 단 0보다 커야 함)
3. `is_active`: 불리언 (선택, 기본값은 True)

---
💡 **[Advanced Tip]**
`price`가 0보다 커야 한다는 조건을 걸기 위해 직접 `if`문으로 검사할 필요가 없습니다. Pydantic에서 제공하는 `Field` 객체(예: `Field(gt=0)`)를 사용하면, 데이터가 들어오는 순간 알아서 에러를 잡아내는 우아한 코드를 작성할 수 있습니다.

```python
from pydantic import BaseModel, Field, ValidationError

# 1. 아래에 Product 모델을 작성하세요.
class Product(BaseModel):
    name: str
    price: int = Field(gt=0)
    is_active: bool = True

# --- [테스트 코드] ---
try:
    # 정상 데이터 테스트
    p1 = Product(name="게이밍 마우스", price=50000)
    print("성공:", p1)
    
    # 에러 데이터 테스트 (가격이 0)
    p2 = Product(name="무료 나눔", price=0)
except ValidationError as e:
    print("검증 에러 발생! 모델 설계가 잘 되었습니다.")
```

## 문제 2: 비동기(Async) 병렬 처리
마치 커피숍에서 3명의 바리스타가 동시에 커피를 내리듯, 파이썬에서 여러 작업을 동시에 처리해주는 비동기 프로그래밍 기초입니다.

`download_file`이라는 1초가 걸리는 비동기 함수가 있습니다.
이 함수를 이용해 총 3개의 파일을 다운로드하려고 합니다. 3개를 순서대로 받으면 3초가 걸리겠지만, **동시에(병렬로) 요청을 던져서 1초 남짓한 시간에 모두 받아오는** `main` 비동기 함수를 완성하세요.

---
💡 **[Advanced Tip]**
`await download_file()`을 세 번 연속으로 적으면 순차 처리(직렬)가 되어버립니다. `asyncio.gather()` 함수를 장바구니처럼 사용해서 3개의 작업을 한 번에 묶어 던지면 진정한 비동기 병렬 처리의 마법이 일어납니다.

```python
import asyncio
import time

async def download_file(file_id: int):
    print(f"파일 {file_id} 다운로드 시작...")
    await asyncio.sleep(1) # 1초 대기 (네트워크 지연 시뮬레이션)
    return f"파일 {file_id} 완료"

async def main():
    start_time = time.time()
    
    # 2. 여기에 비동기 병렬 처리 코드를 작성하세요. (결과를 변수에 담으세요)
    results = await asyncio.gather(download_file(1),download_file(2),download_file(3))
    
    end_time = time.time()
    print("결과:", results)
    print(f"총 걸린 시간: {end_time - start_time:.2f}초")

# 주피터 노트북 환경에서는 아래처럼 실행합니다.
# await main()
```

## 문제 3: FastAPI 라우터 기초 (Path & Query Parameters)
FastAPI에서 클라이언트의 요청 주소를 받아내는 라우터(Endpoint)를 설계해 봅시다.

특정 사용자의 정보를 가져오는 `/users/{user_id}` API를 만들려 합니다.
1. `user_id`는 경로 매개변수(Path Parameter)로 받으며 정수형(int)이어야 합니다.
2. `limit`은 선택적인 쿼리 매개변수(Query Parameter)로 받으며 정수형이고 기본값은 10이어야 합니다. (예: `/users/5?limit=20`)
함수의 반환값으로는 전달받은 `user_id`와 `limit`을 그대로 딕셔너리 형태로 반환되게 구성하세요.

---
💡 **[Advanced Tip]**
FastAPI의 천재적인 부분은 파이썬의 **타입 힌팅(Type Hinting)**을 그대로 활용한다는 점입니다. 함수의 괄호 안에 `user_id: int`와 `limit: int = 10`이라고만 적어주면 프레임워크가 알아서 URL을 분석하고 값을 끼워 넣어 줍니다.

```python
from fastapi import FastAPI

app = FastAPI()

# 3. 아래에 데코레이터와 함수를 작성하세요.
@app.get("/users/{user_id}")
def get_user(user_id: int, limit: int = 10):
    return {"user_id": user_id, "limit": limit}

# (로컬에서 테스트하려면 별도의 파이썬 파일로 저장 후 uvicorn으로 실행해야 합니다)
# 코드 셀에는 구조와 타입 힌팅만 정확히 맞춰주시면 정답으로 인정합니다!
```
