### ë¬¸ì œ 1: ì´ë©”ì¼ ë„ë©”ì¸ ì¶”ì¶œ (String Split)
ìœ ì €ë“¤ì˜ ì´ë©”ì¼ ì£¼ì†Œê°€ ë‹´ê¸´ ë¦¬ìŠ¤íŠ¸ê°€ ìˆìŠµë‹ˆë‹¤.
ê° ì´ë©”ì¼ì—ì„œ `@` ë’¤ì˜ ë„ë©”ì¸ë§Œ ì¶”ì¶œí•˜ì—¬ ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ `domains`ë¥¼ ë§Œë“œì„¸ìš”.

*ì˜ˆì‹œ: 'user@google.com' -> 'google.com'*

```python
emails = [
    "alex@google.com",
    "brian@yahoo.co.kr",
    "chlie@naver.com",
    "david@google.com"
]

domains = []
for info in emails:
    info = info.split('@')
    info = (info[1])
    domains.append(info)


print(domains)
```

### ë¬¸ì œ 3: íŠ¹ì • ë‹¨ì–´ í¬í•¨ ì—¬ë¶€ í•„í„°ë§

### ë¬¸ì œ 4: ê°œì¸ì •ë³´ ë§ˆìŠ¤í‚¹ (String Slicing)
ì£¼ë¯¼ë“±ë¡ë²ˆí˜¸ ë’·ìë¦¬ë¥¼ ê°€ë¦¬ëŠ” í•¨ìˆ˜ `mask_id`ë¥¼ ë§Œë“œì„¸ìš”.
ì…ë ¥ëœ ë¬¸ìì—´ì˜ **ì• 6ìë¦¬ì™€ í•˜ì´í”ˆ(-), ê·¸ë¦¬ê³  ë’·ìë¦¬ ì²« ê¸€ì**ê¹Œì§€ë§Œ ë³´ì—¬ì£¼ê³ ,
ë‚˜ë¨¸ì§€ëŠ” `*`ë¡œ ì±„ì›Œì•¼ í•©ë‹ˆë‹¤.

*ì˜ˆì‹œ: '990101-1234567' -> '990101-1******'*

```python
def mask_id(id_number):
    return id_number[:8] + "*******"

# í…ŒìŠ¤íŠ¸
print(mask_id("990101-1234567"))
print(mask_id("021225-3111111"))
```

### ë¬¸ì œ 5: íˆ¬í‘œ ê²°ê³¼ ì§‘ê³„ (Dictionary Counting)
ì‚¬ëŒë“¤ì´ íˆ¬í‘œí•œ í›„ë³´ì˜ ì´ë¦„ì´ ë¦¬ìŠ¤íŠ¸ì— ë“¤ì–´ìˆìŠµë‹ˆë‹¤.
ê° í›„ë³´ê°€ ëª‡ í‘œë¥¼ ë°›ì•˜ëŠ”ì§€ ë”•ì…”ë„ˆë¦¬ì— ì •ë¦¬í•˜ì„¸ìš”.

*ì¶œë ¥ ì˜ˆì‹œ: {'Alice': 2, 'Bob': 1, 'Charlie': 1}*

```python
votes = ["Alice", "Bob", "Alice", "Charlie", "Alice", "Bob"]

vote_count = {}

for name in votes:
    if name not in vote_count:
        vote_count[name] = 1
    else: 
        vote_count[name] +=1

print(vote_count)
```
