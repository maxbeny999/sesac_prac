### âœ… [ìµœì¢… ì‹¤ìŠµ] ë„ì„œê´€ ì‹œìŠ¤í…œ ì™„ì„± ì½”ë“œ

````python
class Book:
    # statusëŠ” ì…ë ¥ë°›ì§€ ì•Šê³ , ì•ˆì—ì„œ ìë™ìœ¼ë¡œ "available"ë¡œ ì„¤ì •
    def __init__(self, title):
        self.title = title
        self.status = "available"

    def borrow(self):
        self.status = "borrowed"

    def return_book(self):
        self.status = "available"

class Library:
    # books ë¦¬ìŠ¤íŠ¸ëŠ” ì…ë ¥ë°›ì§€ ì•Šê³ , ì•ˆì—ì„œ ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¡œ ìƒì„±
    def __init__(self):
        self.books = []
    
    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, title):
        for book in self.books:
            # 1. ì œëª©ì´ ê°™ê³ (and) 2. ìƒíƒœê°€ ëŒ€ì¶œ ê°€ëŠ¥ì´ë©´
            if book.title == title and book.status == "available":
                book.borrow()
                print(f"'{title}' ëŒ€ì¶œ ì™„ë£Œ")
                return # í•¨ìˆ˜ ì¢…ë£Œ
        
        print("ì±…ì´ ì—†ê±°ë‚˜ ëŒ€ì¶œ ì¤‘ì…ë‹ˆë‹¤.")
````
