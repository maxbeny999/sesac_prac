### 문제 3

### 문제 5
lst = [1, 2, 4, 3, 3, 5, 5, 6, 9, 7] 일 때,
원소가 1, 4, 5, 7인 경우를 제외하고 출력하세요

```python
lst = [1, 2, 4, 3, 3, 5, 5, 6, 9, 7]

for i in lst:
    if i != 1 and i!=4 and i != 5 and i !=7:
        print(i)

exclude_num = [1, 4, 5, 7]

for i in lst:
    if i not in exclude_num:
        print(i)
```

### 문제 6
1 ~ 9 까지의 자연수 중 제곱한 수가 10 이상 50 이하인 자연수의 리스트를 출력해보세요.

```python

lst= []
for i in range(1, 10):
    if 10 <= i ** 2 <= 50:
        lst.append(i)
print(lst)
    

result = [i for i in range(1, 10) if 10 <= i **2 <= 50]
print(result)
```

### 문제 7
숫자 두개를 각각 입력받아 두 수 사이에 속한 모든 정수의 합을 구하시오. (ex. 2와 5를 입력받는다면 2 + 3 + 4 + 5 = 14)

```python
n = int(input())
m = int(input())

lst = []
for i in range(n, m+1):
    lst.append(i)
print(sum(lst))

## 최종 개선판

n = int(input())
m = int(input())

# [핵심 개선] 사용자가 5, 2 처럼 거꾸로 입력해도 2~5로 바꿔줍니다.
# min(n, m)은 작은 수, max(n, m)은 큰 수를 찾아줍니다.
start_num = min(n, m)
end_num = max(n, m)

# [핵심 개선] 굳이 리스트를 만들거나 for문을 돌리지 않아도,
# range 자체를 sum에 넣으면 바로 합계를 구해줍니다.
result = sum(range(start_num, end_num + 1))

print(result)
```
