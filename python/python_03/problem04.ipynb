
### 문제 1
사용자로부터 정수 `n`을 입력받아, 1부터 `n`까지 모든 정수의 합을 출력하세요.

````python
n = int(input())
lst = list(range(1, n+1))

result = sum(lst)

print(result)
````

### 문제 2

지정된 비밀번호(예: `"python123"`)가 입력될 때까지 계속해서 입력을 받는 프로그램을 작성하세요. 틀릴 경우 "다시 시도하세요"를 출력합니다.

````python
while True:
    id = input()
    if id == "python123":
        print("성공")
        break
    else:
        print("다시 시도하세요")
````

### 문제 3

1부터 30까지의 숫자 중에서 3의 배수를 제외한 나머지 숫자들만 출력하세요.

````python
lst = list(range(1, 31))

for no3 in lst:
    if (no3 % 3 != 0):
        print(no3)
````

### 문제 4
`names = ['jun', 'ken', 'alex', 'sophia', 'neo']` 일 때,

출석번호를 포함하여 '1번 jun', ..., '5번 neo'의 형태로 출석을 부르시오.

````python
names = ['jun', 'ken', 'alex', 'sophia', 'neo']


for int, name in enumerate(names, start = 1):
    print(f"{int}번{name}")
````
