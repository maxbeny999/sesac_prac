### 문제 5
기존 `scores` 딕셔너리를 사용하여, 점수가 80점 이상이면 'Pass', 미만이면 'Fail'을 값으로 가지는 새로운 딕셔너리를 만드세요.

```python
# 코드를 여기에 작성하세요
scores = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 95,
    "Eve": 88
}

# 80이상을 담을 리스트 와 80이하를 담을 리스트를 만든다. 반복문을 이용해 문제4 에서 한 것 처럼 점수별로 나눈다.
# 위에서는 나눈학생들의 이름과 점수를  출력했지만 이번에는 이름과 점수에 따라 pass인지 fail인지 입력

#1차 결과물 
over80 = []
under80 = []

for names in scores:
    value = scores[names]
    if value >= 80:
        over80.append(names)
    elif value < 80:
        under80.append(names)
    consequence = {
        f"{over80}" : "Pass",
        f"{under80}" : "Fail"
    }
print(consequence)


##################
#2차 결과물 

consequence = {}

for names in scores:
    value = scores[names]
    if value >= 80:
        consequence[names] = "Pass"
    else :
        consequence[names] = "Fail"
print(consequence)

###################
#ai가 만든 더 짧은 결과물
# {키: 값 for 키, 값 in 딕셔너리.items()}
# 이해하는데 오래걸렸음. 삼항 연산자 사용됨, 영어처럼 어순을 생각하면 자연스럽게 해석 가능함
#  ex 영어 (결과 중심): "Give me(주세요) the menu if(만약) it represents cheaper than 8$ for(대상은) all items."
# 파이썬이 수학적 언어라 가능함,(집합 조건 제시법)
# 일반적 for 문은 어떻게 만들지를 단계별로 설명 
# ex 1. 빈 그릇 생성, 2. 하나 꺼냄, 3. 검사함, 4. 담음.
# 컴프리헨션은 무엇을 만들지 선언하고 만듬(집합 조건 제시법) 
# ex "나는 이런 모양(name: Pass)의 딕셔너리를 원해!" (단, 재료는 scores에서 가져와서...)

consequence = {name: "Pass" if score >= 80 else "Fail" for name, score in scores.items()}

print(consequence)

#################
# 함수 활용 버전

# 1. 채점 기계(함수) 제작
def grading_machine(target_scores):
    result = {} # 빈 딕셔너리 생성
    for name in target_scores:
        value = target_scores[name]
        if value >= 80:
            result[name] = "Pass"
        else:
            result[name] = "Fail"
    return result # 완성된 결과물을 뱉어냄

# 2. 사용 (재활용)
class_A = {"Alice": 85, "Bob": 70, "John" : 100, "Ash": 50}
class_B = {"Charlie": 90, "David": 50, "Kim" : 100, "Ace" : 60}

print(grading_machine(class_A)) # A반 채점해!
print(grading_machine(class_B)) # B반 채점해!
```
