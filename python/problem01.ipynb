### 문제 5
물건의 가격이 price이고, 낸 돈이 money일 때, 거스름돈에서 500원 동전의 개수와 100원 동전의 개수를 구하는 프로그램을 작성하세요. 
(단, 가격은 항상 100원 단위이며 money > price라고 가정합니다.)

예: price = 3200, money = 5000 -> 500원: 3개, 100원: 3개

````python
price = 3200
money = 5000
coin500 = 500 
coin100 = 100

change = money-price
leftover = change % coin500

print(change) # 거스름돈
print(change // coin500) # 500원의 개수
print(change % coin500) # 500원 내고 남은 돈
print(leftover // coin100) # 100원의 개수
````
