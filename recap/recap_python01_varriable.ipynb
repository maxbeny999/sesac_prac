# 변수와 할당
 - 변수는 데이터 값을 정하는 상자. EX number = 100 **변수 number에 100이란 값이 들어있음을 뜻한다**
 - 할당, 변수를 생성할 때는 **할당 연산자(=)을 사용한다 변수 이름 = 값 EX number(변수 이름) = 100(값)
 - 수학의 등호(=) 처럼 **작동하지 않는다**. 왼쪽과 오른쪽이 같다는 의미가 아님, **왼쪽 변수에 오른쪽 값이 들어가 있다는 의미** 

```python
number = 100 

name = "max" # 문자열 = str
age = 25 # 정수형 = int
is_male = True # 불린형 = True or False

print(name, age, is_male)
print(type(name), type(age), type(is_male))
```
