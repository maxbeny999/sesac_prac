🔥 Level 1. 기초 체력 (변수, 입력, 출력)
시나리오: 당신은 챗봇을 처음 만드는 개발자입니다. 사용자에게 정보를 받아 깔끔하게 정리해주는 'AI 비서'의 기초를 다져보세요.
문제:
사용자에게 이름을 입력받으세요.
사용자에게 태어난 연도(예: 1990)를 입력받으세요.
현재 연도(2025년 기준)를 변수에 저장해두고, 사용자의 한국 나이를 계산하세요.
아래와 같은 형식으로 출력하세요.
출력 예시:
"[AI] 안녕하세요 kyle님! 2025년 기준, 당신은 36세이시군요."
사용 문법: input(), int() 형변환, 산술 연산, print() (f-string 추천)
💡 힌트: input으로 받은 년도는 문자열입니다. 계산하려면 숫자로 바꿔야겠죠?
🔥 Level 2. 판단력 기르기 (조건문, 논리 연산)
시나리오: 사용자의 기분에 따라 저녁 메뉴를 추천해주는 '메뉴 추천 봇'을 만들어봅니다.
문제:
사용자에게 "지금 기분이 어떠신가요? (좋음/나쁨/그저그럼)"을 입력받으세요.
if문을 사용하여 다음 조건대로 추천 메뉴를 출력하세요.
좋음: "치킨에 맥주 어떠세요?"
나쁨: "매운 떡볶이로 스트레스를 날려보세요."
그저그럼: "따뜻한 국밥이 좋겠네요."
그 외 다른 입력: "알 수 없는 기분이네요. 물 한 잔 드세요."
(심화) 기분이 "나쁨"이거나 "그저그럼"일 때는 메뉴 추천 뒤에 "힘내세요!"라는 응원 문구도 추가로 출력해보세요. (or 연산자 활용)
사용 문법: if, elif, else, ==, or
🔥 Level 3. 데이터 창고 관리 (리스트 기초)
시나리오: 당신은 '할 일 관리(To-Do List)' 에이전트를 개발 중입니다. 사용자의 할 일을 저장하고 관리하는 로직을 짜보세요.
문제:
비어 있는 리스트 todos를 만드세요.
todos에 "파이썬 공부", "운동하기", "독서" 세 가지를 순서대로 추가하세요(append).
사용자가 "운동하기"를 완료했다고 가정하고, 리스트에서 삭제하세요(remove).
갑자기 급한 일인 "저녁 약속"이 생겼습니다. 리스트의 맨 앞에 끼워 넣으세요(insert).
최종 리스트의 내용과, 총 할 일이 몇 개인지(len) 출력하세요.
사용 문법: [], .append(), .remove(), .insert(), len()
🔥 Level 4. 자동화의 시작 (반복문 - for)
시나리오: 서버에서 수집된 데이터에 오류(음수)가 섞여 있습니다. 정상적인 데이터만 골라내는 '데이터 필터링' 시스템을 만드세요.
문제:
data = [10, -5, 20, -1, 50, 30] 라는 리스트가 있습니다.
for 문을 사용해서 리스트의 숫자를 하나씩 꺼내세요.
꺼낸 숫자가 0보다 작으면(음수), "오류 데이터 발견: (숫자)"를 출력하고 건너뛰세요(continue).
정상 데이터(양수)라면 "정상 처리: (숫자)"를 출력하세요.
(심화) 반복문이 끝난 후 "데이터 처리가 완료되었습니다."를 출력하세요.
사용 문법: for, if, continue, 비교 연산
🔥 Level 5. 검색 엔진 만들기 (종합 - while, break, in)
시나리오: 사용자가 찾고 싶은 단어를 입력하면, 가지고 있는 데이터베이스에서 검색해주는 '검색 에이전트'입니다.
문제:
database = ["AI", "Python", "Agent", "LangChain", "Chatbot"] 리스트를 만드세요.
while True:를 써서 무한히 입력을 받으세요.
사용자에게 "검색할 키워드를 입력하세요 (종료는 'q'): "라고 물어보세요.
조건 1: 입력값이 'q'이면 "프로그램을 종료합니다." 출력 후 반복문을 탈출(break)하세요.
조건 2: 입력한 키워드가 database 안에 있으면(in), "데이터베이스에 존재합니다!" 출력.
조건 3: 없으면, "찾는 데이터가 없습니다." 출력.
사용 문법: list, while, input, if, break, in 연산자
