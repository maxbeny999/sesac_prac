# 시나리오: 놀이공원 입장 시스템

    promblem02 문제7번 추가 연습용

1. **키(height)**가 120cm 이상이어야 합니다.
2. 그리고(and), **나이(age)**가 15세 이상이거나 또는(or) 보호자 동반 여부(with_parent)가 "yes"여야 합니다.
3. 단(and), 이름(name)이 출입금지 명단인 "Voldemort"와 같지 않아야(!=) 합니다.

```python
height = int(input())
age = int(input())
name = input()

with_parents = (input("yes / no") == "yes")

if (height >= 120 and name != "Voldemort") and (age >= 15 or with_parents == "yes"):
    print("놀이공원에 입장 할 수 있습니다.")
else:
    print("놀이공원에 입장 할 수 없습니다.")
```

```python
# 1. 안내 문구와 함께 입력 받기
print("--- 놀이공원 입장 시스템 ---")
height = int(input("키를 입력하세요(cm): "))
age = int(input("나이를 입력하세요: "))
name = input("이름을 입력하세요: ")

# 2. 입력값 전처리 (공백 제거 + 소문자 변환)
# " Yes " -> "yes"로 바꿔서 비교하므로 에러가 줄어듭니다.
input_parent = input("보호자가 있나요? (yes/no): ").strip().lower()
has_parent = (input_parent == "yes")

# 3. 조건의 의미를 변수로 정리하기 (가독성 UP)
# '조건식' 자체가 아니라 '조건의 의미'가 보입니다.
is_banned = (name == "Voldemort")      # 입장이 금지된 사람인가?
is_tall_enough = (height >= 120)       # 키가 충분한가?
can_pass_alone = (age >= 15)           # 혼자 탈 수 있는 나이인가?

# 4. 최종 판단 (영어 문장처럼 읽힘)
# "금지된 사람이 아니고, 키가 충분하며, (어른이거나 보호자가 있거나)"
if not is_banned and is_tall_enough and (can_pass_alone or has_parent):
    print(f"환영합니다, {name}님! 입장 가능합니다.")
else:
    print(f"죄송합니다, {name}님. 입장할 수 없습니다.")
```

더 완벽한코드

1. .strip().lower():
사용자가 실수로 스페이스바를 눌러 " yes"라고 치거나, 대문자로 "YES"라고 쳐도 찰떡같이 알아듣습니다. AI 서비스를 만들 때 사용자 입력을 정제하는 아주 중요한 습관입니다.
변수명으로 로직 분리 (is_banned, is_tall_enough ...):

2. 나중에 "입장 금지 대상"이 바뀌거나 "키 제한"이 바뀌면, 아래 if문은 건드릴 필요 없이 위쪽 변수만 살짝 고치면 됩니다. 유지보수가 훨씬 편해집니다.

3. not is_banned:
name != "Voldemort" 보다 not is_banned (밴 당한 사람이 아니면) 라고 읽는 게 훨씬 직관적입니다.

이런 식으로 **"코드를 읽는 사람(미래의 나, 동료)을 배려하는 작성법"** 을 익히시면 실력이 금방 늡니다!
