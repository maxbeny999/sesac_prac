```python
import requests
from pprint import pprint
import os
from dotenv import load_dotenv

load_dotenv()

URL = "https://api.themoviedb.org/3/movie/now_playing"
API_KEY = os.getenv('TMDB_API_KEY')

params = {
    'language' : 'ko-kr',
    # 'api_key' : "짧은 그 키"
}

headers = {
    "Authorization" : f"Bearer {API_KEY}"
}

try:
    total_data = []
    
    for page_num in range(1, 2):
        params['page'] = page_num

        response = requests.get(URL, headers=headers, params=params)

        response.raise_for_status()
    # if not response.status_code == 200:
    #     raise Exception
    
        data = response.json()

        result = data['results']

        total_data.extend(result)

    pprint(total_data)
    print(len(total_data))
    
except Exception as e:
    print(e)




# except requests.exceptions.HTTPError as errh:
#     print(f"Http Error: {errh}") # 401(인증오류), 404(경로오류) 등
# except requests.exceptions.ConnectionError as errc:
#     print(f"Error Connecting: {errc}") # 인터넷 연결 끊김
# except requests.exceptions.Timeout as errt:
#     print(f"Timeout Error: {errt}") # 서버 응답 지연
# except requests.exceptions.RequestException as err:
#     print(f"{err}") # 그 외 requests 관련 에러
# except ValueError:
#     print("JSON Decoding failed") # 응답이 JSON 형식이 아닐 때
# except Exception as e:
#     print(f"Unknown Error: {e}") # 정말 예상치 못한 일반 에러
```

### 문제 1
현재 상영 중인 영화(Now Playing) 중 평점(vote_average)이 가장 높은 영화를 찾으시오.

```python
import requests

try:
    response = requests.get(URL, headers=headers, params=params)
    response.raise_for_status()
    data = response.json()
    movie_list = data['results']

    max_vote = 0
    best_movie = None

    for movie in movie_list:
        vote_average = movie['vote_average']
        
        if vote_average > max_vote:
            max_vote = vote_average
            best_movie = movie
    if best_movie:
        print(f"최고 평점 영화: {best_movie['title']}")
        print(f"평점: {best_movie['vote_average']}")
    else:
        print("영화 데이터가 없습니다.")

except Exception as e:
    print(e)
```

### 문제 3
현재 상영 중인 영화(now_playing) 데이터에서 다음 정보만 담긴 리스트를 만드세요.

- title (영화 제목)  
- vote_average (평점)  
- release_date (개봉일)

```python
import requests
from pprint import pprint

URL = "https://api.themoviedb.org/3/movie/now_playing"


try:
    response = requests.get(URL, headers=headers, params=params)
    response.raise_for_status()
    data = response.json()
    data = data['results']

    # items 를 쓰고 싶음 items는 딕셔너리에만 됨
    # data['results'] 를 반복문을 돌려서 리스트 안의 딕셔너리를 하나씩 뽑는다. 그러면 리스트를 뜯어서 내부의 딕셔너리들을 하나씩 반복문에서 인식하게 해서 
    # - title (영화 제목)  
    # - vote_average (평점)  
    # - release_date (개봉일)
    # 를 items() 를 통해서 제목, 평점, 개봉일만 내가 원하는것을 뽑을 수 있다.

    # items 필요없이 그냥 딕셔너리명['키값'] 으로 충분했다.

    final_list = []

    for movie_list in data:
        final_list.append(movie_list['title'])
        final_list.append(movie_list['vote_average'])
        final_list.append(movie_list['release_date'])
    pprint(final_list)
        
        

except Exception as e:
    pprint(e)
```

### 문제 4
현재 상영 중인 영화(now_playing) 데이터를 vote_average(평점)순으로 정렬하여 영화 제목이 담긴 리스트를 만드세요.
단, 리스트의 길이는 100 이상이어야 합니다.


```python
import requests
from pprint import pprint

URL = "https://api.themoviedb.org/3/movie/now_playing"


params = {
    'language' : 'ko-kr',
    # 'api_key' : "짧은 그 키"
    'page' : 1,
}

headers = {
    "Authorization" : f"Bearer {API_KEY}"
}

try:
    total_data = [] 

    for page_num in range(1, 6):
        params['page'] = page_num

        response = requests.get(URL, headers=headers, params=params)
        response.raise_for_status()
        data = response.json()
        results = data['results']

        total_data.extend(results)
    # sort, sorted 처음 써봄. lambda는 뭔지는 알았지만 제대로 활용한 건 처음.
    # lambda 는 임시함수. 간단한 함수를 귀찮게 def ***() 이렇게 만들지 않고 key = lambda x : x 이렇게 때웠음
    sort_byAve = sorted(total_data, key=lambda el: el['vote_average'], reverse=True  ) 
    final_list = []
    

    for movie_list in sort_byAve:
        final_list.append(movie_list['title'])

    pprint(final_list)
    print(len(final_list))

except Exception as e:
    pprint(e)
```
